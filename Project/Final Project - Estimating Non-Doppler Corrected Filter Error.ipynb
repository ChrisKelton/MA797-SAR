The main goal of this project is to estimate the error in the case where the Generalized Ambigutity Function (GAF) filter has not been corrected for Doppler and analyze our results following the guidelines in Section 6.6.5 in [1]. We will derive the GAF without factorization and has been doppler corrected and, subsequently, the factorized GAF without doppler correction and estimate the error between the two.

The goal of this paper is to derive expression [1, (eq. 6.109)],

$$\frac{max\left|W - W_{(R, \sum)} \right|}{max\left|W_{(R, \sum)} \right|} \lesssim \frac{\pi}{8}\frac{B}{\omega_{0}}\cdot const,$$

and analyze the results. Where $W$ is the non-factorized, doppler-corrected with respect to the target GAF, $W_{(R, \sum)}$ is the factorized, doppler-corrected with respect to the target GAF, $B$ is the bandwidth, $\omega_{0}$ is the carrier frequency, and $const$ will be a constant with some assumptions regarding the scale of factors, which we will also further explore when certain values exceed our assumption that they are of the order $\mathcal{O(1)}$.

We will begin by determining the general form of the GAF that is non-factorized and doppler corrected, where we will see that the non-doppler corrected form is easily extracted from the doppler corrected form. Since we are taking into account the movement of the platform during the transmission of the radar signals emitted, we will utilize the Lorentz transform in order to transform a moving object to a coordinate system that is stationary, while keeping the wave equation and Maxwell's equations invariant. Thus, we will have $\left(t, z_{1}, z_{2}, z_{3} \right) \mapsto \left(\sigma, \zeta_{1}, z_{2}, z_{3} \right)$. Where,

$$\sigma = \frac{1}{\beta}\left(t - \frac{vz_{1}}{c^{2}} \right), \quad \zeta_{1} = \frac{1}{\beta}\left(-vt + z_{1} \right),$$
and
$$\beta = \sqrt{1 - \frac{v^{2}}{c^{2}}}.$$

Thus, leading to the following wave equation:

$$\begin{split} \frac{1}{c^{2}}\frac{\partial^{2}u}{\partial\sigma^{2}} - \frac{\partial^{2}u}{\partial\zeta_{1}^{2}} - \frac{\partial^{2}u}{\partial z_{2}^{2}} - \frac{\partial^{2}u}{\partial z_{3}^{2}} &= P\left(\frac{\sigma}{\beta} + \frac{\frac{v}{c}}{\beta}\frac{\zeta_{1}}{c} \right)\delta\left(\beta\zeta_{1} \right)\delta\left(z_{2} + L \right)\delta\left(z_{3} - H \right) \\
&= f\left(\sigma, \zeta_{1}, z_{2}, z_{3} \right). \end{split}$$

Thus, using Kirchhoff's integral theorem to define the impinging field, we get

$$u^{(0)}\left(\sigma, \zeta_{1}, z_{2}, z_{3} \right) = \frac{1}{4\pi}\int\int\int\frac{f\left(\sigma - \frac{\rho'}{c}, \zeta_{1}', z_{2}', z_{3}' \right)}{\rho'}d\zeta_{1}'dz_{2}'dz_{3}',$$
where
$$\rho' = \left(\left(\zeta_{1} - \zeta_{1}' \right)^{2} + \left(z_{2} - z_{2}' \right)^{2} + \left(z_{3} - z_{3}' \right)^{2} \right)^{\frac{1}{2}}.$$

Thus, defining $\zeta_{1}' = 0$, as we have transformed into a coordinate system where our antenna is no longer moving, $z_{2}' = -L$, as this is the projection distance of the antenna, $z_{3}' = H$, as this is the height of the antenna from the ground. Thus, we are saying that the antenna is at rest at $\boldsymbol{x} = (0, -L, H)$. Integrating $u^{(0)}$ we get,

$$u^{(0)} = \frac{1}{4\pi\beta}\frac{P\left(\frac{\sigma - \frac{\rho}{c}}{\beta} \right)}{\rho},$$
where
$$\rho = \left(\zeta_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2} \right)^{\frac{1}{2}}.$$

Notice, if we drop $\beta$, then we get the straightforward retarded potential in the new coordinate system. Solving for $\rho$,

$$\rho = \left(\zeta_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2} \right)^{\frac{1}{2}}$$

$$\begin{split} \zeta_{1}^{2} &= \left(\frac{1}{\beta}\left(-vt + z_{1} \right) \right)^{2} \\
&= \frac{1}{\beta^{2}}\left(z_{1} - vt \right)^{2} \\
&= \frac{1}{\beta^{2}}\left(z_{1}^{2} - 2z_{1}vt + \left(vt \right)^{2} \right) \\
&= \frac{z_{1}^{2}}{\beta^{2}} + \frac{1}{\beta^{2}}\left(-2z_{1}vt + \left(vt \right)^{2} \right) \end{split},$$
where
$$\begin{split} \frac{z_{1}^{2}}{\beta^{2}} &= z_{1}^{2} - z_{1}^{2} + \frac{z_{1}^{2}}{\beta^{2}} \\
&= z_{1}^{2} + z_{1}^{2}\left(\frac{1}{\beta^{2}} - 1 \right) \\
&= z_{1}^{2} + \frac{1 - \beta^{2}}{\beta^{2}}z_{1}^{2} \\
&= z_{1}^{2}. \end{split}$$

Where we dropped any $\frac{c^{2}}{v^{2}} \gg 1$ terms. Thus,

$$\begin{split} \zeta_{1}^{2} &= z_{1}^{2} + \frac{1}{\beta^{2}}\left(-2z_{1}vt + \left(vt \right)^{2} \right) \\
&= z_{1}^{2} - 2z_{1}vt + \left(vt \right)^{2} + \frac{1 - \beta^{2}}{\beta^{2}}\left(z_{1} - vt \right)^{2} \\
&= z_{1}^{2} - 2z_{1}vt + \left(vt \right)^{2} \end{split}$$

$$\rho \approx \left(z_{1}^{2} - 2z_{1}vt + \left(vt \right)^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2} \right)^{\frac{1}{2}}$$

$$r = R_{z}\bigg\rvert_{t=0,z_{3}=0} = \sqrt{z_{1}^{2} + \left(z_{2} + L \right)^{2} + H^{2}}.$$

Where $r$ defines the moment of time $t = 0$ when the center of the pulse is emitted corresponds to the location $\boldsymbol{x}(0) = (0, -L, H)$ on the orbit. Therefore, when $z_{3} = 0$,

$$\rho \approx \left(r^{2} - 2z_{1}vt + \left(vt \right)^{2} \right)^{\frac{1}{2}} .$$

Furthermore, the timescale we are interested in is the travel time of the impinging signal to propagate to the target or $t \sim \frac{r}{c}$. Therefore, the 3rd term on the right-hand side is only taken into account if we have a very narrow sub-angle or $\left|z_{1}\right|\left|vt\right| \sim \left(vt \right)^{2}$ as $\left(vt \right)^{2} = \left(v\frac{r}{c} \right)^{2} = \left(\frac{v}{c}r \right)^{2} = \frac{v^{2}}{c^{2}}r^{2}$, so we would assume to drop this term as well due to $\frac{v^{2}}{c^{2}}$. Henceforth,

$$\begin{split} \rho &\approx \sqrt{r^{2} - 2z_{1}vt} \\
&= r\sqrt{1 - \frac{2z_{1}vt}{r^{2}}} \\
&\approx r\left(1 - \frac{1}{2}\frac{2z_{1}vt}{r^{2}} + \frac{1}{4}\frac{4z_{1}^{2}\left(vt \right)^{2}}{r^{4}} \right) \\
&= r - \frac{z_{1}vt}{r}. \end{split}$$

Substituting this into $u^{(0)}$ and dropping the $\beta \sim 1$ terms,

$$\begin{split} u^{(0)}(t, \boldsymbol{z}) &\approx \frac{1}{4\pi}\frac{P\left(\sigma - \frac{\rho}{c} \right)}{\rho} \\
&\approx \frac{1}{4\pi}\frac{P\left(t - \frac{vz_{1}}{c^{2}} - \frac{r}{c} - \frac{v}{c}\frac{z_{1}t}{r} \right)}{\rho} \\
&= \frac{1}{4\pi}\frac{P\left(t\left(1 - \frac{v}{c}\frac{z_{1}}{r} \right) - \frac{vz_{1}}{c^{2}} - \frac{r}{c} \right)}{\rho} \\
&= \frac{1}{4\pi}\frac{P\left(\left(t - \frac{r}{c} \right)\left(1 + \frac{v}{c}\frac{z_{1}}{r} \right) \right)}{\rho}. \end{split}$$

Integrating $\rho$ with respect to $t$, we get $\left|\frac{\partial\rho}{\partial t} \right| \approx \frac{z_{1}v}{r}$ and since we are not considering a narrow sub-angle this will result in a small amplitude and we can simply replace the slowly oscillating $\rho$ with $r$. Thus,

$$u^{(0)}(t, \boldsymbol{z}) \approx \frac{1}{4\pi}\frac{P\left(\left(t - \frac{r}{c} \right)\left(1 + \frac{v}{c}\frac{z_{1}}{r} \right) \right)}{r},$$

and $\frac{z_{1}}{r}$ defines the look angle of the satellite, which, by the law of cosines, will replace with $\cos(\gamma_{z}(0))$. Where $\gamma_{z}$ is constantly changing as the satellite is moving, however, we will only take the moment of emission at $t = 0$, as, from previous derivations, conclude that the change in range position of the satellite is very small. We will denote $\cos(\gamma_{z}(0)) = \cos(\gamma_{z})$. Therefore,

$$u^{(0)}(t, \boldsymbol{z}) \approx \frac{1}{4\pi}\frac{P\left(\left(t - \frac{r}{c} \right)\left(1 + \frac{v}{c}\cos(\gamma_{z}) \right)\right)}{r}.$$

Then, the receiving signal, $u^{(1)}$ will become:

$$u^{(1)}(t, \boldsymbol{x}') = \int\int\int\nu(z)P\left(\left(t - \frac{r}{c} - \frac{\left|\boldsymbol{x}' - \boldsymbol{z} \right|}{c} \right)\left(1 + \frac{v}{c}cos(\gamma_{z}) \right) \right)d\boldsymbol{z}.$$

Note, if $v = 0$, we get back the solution using the Start-Stop approximation. Now, we need to perform another Lorentz transform on the received, scattered, field at the moving antenna receiver. Therefore, we must perform the mapping $(t, x_{1}', x_{2}', x_{3}') \mapsto (\vartheta, \xi_{1}, x_{2}', x_{3}')$, where

$$\vartheta = \frac{1}{\beta}\left(t - \frac{vx_{1}'}{c^{2}} \right), \quad \xi_{1} = \frac{1}{\beta}\left(-vt + x_{1}' \right).$$

Note, $t - \frac{r}{c}$ is the moment the pulse is reflected at the target, defining $\breve{t} = \frac{r}{c}$. Therefore, the moment the emitting pulse is re-emitted at the target $\boldsymbol{z}$, event of reflection, is

$$\breve{\vartheta} = \frac{1}{\beta}\left(\breve{t} - \frac{vz_{1}}{c} \right), \quad \breve{\xi}_{1} = \frac{1}{\beta}\left(-v\breve{t} + z_{1} \right).$$

We also know, from derivations in earlier homeworks, that the maximum of the $sinc$ argument in the impinging field, $u^{(0)}\left(t, \boldsymbol{z}\right),$ definition is $\frac{\lambda_{0}}{D}.$ Where $\lambda_{0}$ is the wavelength of the emitting radiation pattern and $D$ is the length the of synthetic aperture. Thus the upper bound on the emitting angle is $\frac{\lambda_{0}}{D}$, where this produces the maximum level of radiation emitted by a linear antenna, when parallel to the orbit. Henceforth, the condition where the moment of reflection of the point $\boldsymbol{z}$ is within the antenna beam is expressed as:

$$\frac{\left|\breve{\xi_{1}} \right|}{R} \leq \frac{\lambda_{0}}{D}.$$

For $t \gt \breve{t} = \frac{r}{c}$, the scattered field is obtained by:

$$u^{(1)}\left(t, \boldsymbol{x}'\right) \approx \int \nu(\boldsymbol{z}) P\left(\left(t - \breve{t} - \frac{r'}{c} \right) \left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{\beta R} \right)d\boldsymbol{z}.$$

Where $t - \breve{t} - \frac{r'}{c}$ comes from the event of reflection at the target, $\breve{t},$ and the modified distance between the moving antenna, $\boldsymbol{x}'$, and the target, $\boldsymbol{z}.$ Which is a doppler shift of the retarded time. Where,

$$\begin{split} r' &= \sqrt{\left(x_{1}' - z_{1} \right)^{2} + \left(z_{2} + L \right)^{2} + H^{2}} \\
&= \left|\boldsymbol{x}' - \boldsymbol{z} \right|. \end{split}$$

The indicator function, $\chi_{\Theta}$ comes from $\frac{\left|\breve{\xi}_{1} \right|}{R} \leq \frac{\lambda_{0}}{D},$ where this is the permissible range of the angle of the emitting radiation.

Performing the inverse transform of $\vartheta$ & $\xi_{1}$, we get the following expressions:

$$t = \frac{1}{\beta}\left(\vartheta + \frac{v\xi_{1}}{c^{2}} \right), \quad x_{1}' = \frac{1}{\beta}\left(v\vartheta + \xi_{1} \right).$$

Thus, substituting all values into $t - \breve{t} - \frac{r'}{c}$:

$$t - \breve{t} - \frac{r'}{c} = \frac{1}{\beta}\left(\vartheta + \frac{v\xi_{1}}{c^{2}} \right) - \frac{1}{\beta}\left(\breve{\vartheta} + \frac{c\breve{\xi_{1}}}{c^{2}} \right) - \frac{1}{c}\sqrt{\left(\frac{1}{\beta}\left(v\vartheta + \xi_{1} \right) - z_{1} \right)^{2} + \left(z_{2} + L \right)^{2} + H^{2}}.$$

Then, taking $\beta \approx 1$ and finding the inverse of $z_{1}$ with regards to $\breve{\vartheta}$ & $\breve{\xi_{1}}$, we get

$$t - \breve{t} - \frac{r'}{c} \approx \left(\vartheta - \breve{\vartheta} \right) + \frac{v}{c^{2}}\left(\xi_{1} - \breve{\xi_{1}} \right) - \frac{1}{c}\sqrt{\left(\xi_{1} - \left(\breve{\xi_{1}} + v\breve{\vartheta} \right) + v\vartheta \right)^{2} + \left(z_{2} + L \right)^{2} + H^{2}}.$$

Solving for $\left(\xi_{1} - \left(\breve{\xi_{1}} + v\breve{\vartheta} \right) + v\vartheta \right)^{2}$:

$$\begin{split} \left(\xi_{1} - \left(\breve{\xi_{1}} + v\breve{\vartheta} \right) + v\vartheta \right)^{2} &= \left(\xi_{1} - \breve{\xi_{1}} - v\breve{\vartheta} + v\vartheta \right)^{2} \\
&= \xi_{1}^{2} - \xi_{1}\breve{\xi_{1}} - \xi_{1}v\breve{\vartheta} + \xi_{1}v\vartheta - \xi_{1}\breve{\xi_{1}} + \breve{\xi_{1}}^{2} \\
&+ \breve{\xi_{1}}v\breve{\vartheta} - \breve{\xi_{1}}v\vartheta - \xi_{1}v\breve{\vartheta} + \breve{\xi_{1}}v\breve{\vartheta} + \left(v\breve{\vartheta} \right)^{2} - v^{2}\breve{\vartheta}\vartheta + \xi_{1}v\vartheta - \breve{\xi_{1}}v\vartheta - v^{2}\breve{\vartheta}\vartheta + \left(v\vartheta \right)^{2} \\
&= \left(\xi_{1}^{2} - 2\xi_{1}\breve{\xi_{1}} + \breve{\xi_{1}}^{2} \right) + \left(2\xi_{1}v\vartheta - 2\breve{\xi_{1}}v\vartheta - 2\xi_{1}v\breve{\vartheta} + 2\breve{\xi_{1}}v\breve{\vartheta} \right) + \left(\left(v\vartheta \right)^{2} - 2v^{2}\vartheta\breve{\vartheta} \right) \\
&= \left(\xi_{1} - \breve{\xi_{1}} \right)^{2} + 2v\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right) + v^{2}\left(\vartheta - \breve{\vartheta} \right)^{2} \end{split}.$$

Therefore,

$$t - \breve{t} - \frac{r'}{c} = \left(\vartheta - \breve{\vartheta} \right) + \frac{v}{c^{2}}\left(\xi_{1} - \breve{\xi_{1}} \right) - \frac{1}{c}\sqrt{\left(\xi_{1} - \breve{\xi_{1}} \right)^{2} + 2v\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right) + v^{2}\left(\vartheta - \breve{\vartheta} \right)^{2} + \left(z_{2} + L \right)^{2} + H^{2}}.$$

Introducing, $\varrho = \sqrt{\left(\xi_{1} - \breve{\xi_{1}} \right)^{2} + \left(z_{2} + L \right)^{2} + H^{2} }$

$$t - \breve{t} - \frac{r'}{c} = \left(\vartheta - \breve{\vartheta} \right) + \frac{v}{c^{2}}\left(\xi_{1} - \breve{\xi_{1}} \right) - \frac{1}{c}\sqrt{\varrho^{2} + 2v\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right) + v^{2}\left(\vartheta - \breve{\vartheta} \right)^{2} }.$$

Analyzing the third term containing the square root,

$$\frac{1}{c}\sqrt{\varrho^{2} + 2v\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right) + v^{2}\left(\vartheta - \breve{\vartheta} \right)^{2}} = \sqrt{\frac{\varrho^{2}}{c^{2}} + \frac{2v}{c^{2}}\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right) + \frac{v^{2}}{c^{2}}\left(\vartheta - \breve{\vartheta} \right)^{2} }.$$

The $\frac{v^{2}}{c^{2}}$ term is approximately zero. Therefore,

$$\frac{1}{c}\sqrt{\varrho^{2} + 2v\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right) + v^{2}\left(\vartheta - \breve{\vartheta} \right)^{2}} \approx \frac{1}{c}\sqrt{\varrho^{2} + 2v\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right)}.$$

Applying the Taylor series to the expression under the square root where $\alpha = \frac{2v}{\varrho^{2}}\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right)$ and is sufficiently small, so $\sqrt{1 + \alpha} \approx 1 + \frac{1}{2}\alpha - \frac{1}{4}\alpha^{2} + \cdots$ and taking the first two terms:

$$\begin{split} \frac{1}{c}\sqrt{\varrho^{2} + 2v\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right) + v^{2}\left(\vartheta - \breve{\vartheta} \right)^{2}} &\approx \frac{\varrho}{c} + \frac{\varrho}{c}\frac{1}{2}\left(\frac{2v}{\varrho^{2}}\left(\vartheta - \breve{\vartheta} \right) \left(\xi_{1} - \breve{\xi_{1}} \right) \right) \\
&= \frac{\varrho}{c} + v\left(\vartheta - \breve{\vartheta} \right)\frac{1}{c}\frac{\xi_{1} - \breve{\xi_{1}}}{\varrho}. \end{split}$$

Thus,

$$\begin{split} t - \breve{t} - \frac{r'}{c} &\approx \left(\vartheta - \breve{\vartheta} \right) + \frac{v}{c^{2}}\left(\xi_{1} - \breve{\xi_{1}} \right) - \frac{\varrho}{c} - v\left(\vartheta - \breve{\vartheta} \right)\frac{1}{c}\frac{\xi_{1} - \breve{\xi_{1}}}{\varrho} \\
&= \left(\vartheta - \breve{\vartheta} - \frac{\varrho}{c} \right)\left(1 + \frac{v}{c}\frac{\breve{\xi_{1}} - \xi_{1}}{\varrho} \right). \end{split}$$

The quantity $\frac{\breve{\xi_{1}} - \xi_{1}}{\varrho}$ can be interpreted as cosine of the angle between the velocity and the direction of the observation point, where the scattered field is received on the antenna, to the pulse re-emission location $\breve{\xi_{1}}$ at the re-emission time $\breve{\vartheta}$, thus,

$$\frac{\breve{\xi_{1}} - \xi_{1}}{\varrho} = \cos(\gamma_{\boldsymbol{z}}').$$

Therefore,

$$t - \breve{t} - \frac{r'}{c} = \left(\vartheta - \breve{\vartheta} - \frac{\varrho}{c} \right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}') \right).$$

Thus, substituting the above expression into the defintion of $u^{(1)}$ we get:

$$u^{(1)}\left(t, \boldsymbol{x}' \right) \approx \int\nu(\boldsymbol{z})P\left(\left(\vartheta - \breve{\vartheta} - \frac{\varrho}{c} \right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}') \right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right)d\boldsymbol{z}.$$

The overall Doppler effect for the scattered field received at the antenna is a combination of two contributions. One, where the field is emitted by a moving antenna onto a motionless target, which is given by $1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}})$. The second, where the impinging field is scattered by a motionless target onto a moving antenna, which is given by $1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}')$. Expanding our expression for  $u^{(1)}$ and neglecting the $\frac{v^{2}}{c^{2}}$ terms, we get:

$$u^{(1)}(t, \boldsymbol{x}') \approx \int \nu(\boldsymbol{z})P\left(\left(\vartheta - \breve{\vartheta} - \frac{\varrho}{c} \right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}') + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right)d\boldsymbol{z}.$$

This expression of the scattered field is still in the Lorentz-transformed coordinates, so we will want to convert the coordinate system back into our original cartesian coordinate system,

$$u^{(1)}(t, \boldsymbol{x}') \approx \int \nu(\boldsymbol{z})P\left(\left(t - \frac{vx_{1}'}{c^{2}} - \frac{r}{c} + \frac{vz_{1}}{c^{2}} - \frac{\varrho}{c} \right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}') + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right)d\boldsymbol{z}.$$

When the scattered signal is received at the antenna, $\xi_{1} = 0$, therefore,

$$\begin{split} \varrho &= \sqrt{\breve{\xi_{1}}^{2} + \left(z_{2} + L \right)^{2} + H^{2}} \\
&= \sqrt{\left(z_{1} - \frac{v}{c}r \right)^{2} + \left(z_{2} + L \right)^{2} + H^{2}} \\
&= \sqrt{z_{1}^{2} - 2\frac{v}{c}z_{1}r + \frac{v^{2}}{c^{2}}r^{2} + \left(z_{2} + L \right)^{2} + H^{2}} \\
&= \sqrt{z_{1}^{2} - 2\frac{v}{c}z_{1}r + \left(z_{2} + L \right)^{2} + H^{2}} \\
&= \sqrt{r^{2} - 2z_{1}\frac{v}{c}r} \\
&= r\sqrt{1 - 2z_{1}\frac{v}{cr}} \\
&\approx r\left(1 - \frac{1}{2}\left(2z_{1}\frac{v}{cr} \right) + \frac{1}{4}\left(2z_{1}\frac{v}{cr} \right)^{2} \right) \\
&\approx r - z_{1}\frac{v}{c}, \end{split}$$

where we neglected the $\frac{v^{2}}{c^{2}}$ terms. Substituting this expression into $\cos(\gamma_{\boldsymbol{z}}')$ at $\xi_{1} = 0$ and the inverse of $\breve{\xi_{1}}$, we get

$$\begin{split} \cos(\gamma_{\boldsymbol{z}}') = \frac{\breve{\xi_{1}} - \xi_{1}}{\varrho}\bigg\rvert_{\xi_{1} = 0} &\approx \frac{z_{1} - \frac{v}{c}r}{r - z_{1}\frac{v}{c}} \\
&\approx \frac{z_{1} - \frac{v}{c}r}{r\left(1 - \frac{z_{1}}{r}\frac{v}{c} \right)} \\
&\approx \frac{z_{1} - \frac{v}{c}r}{r}\left(1 + \frac{z_{1}}{r}\frac{v}{c} \right) \\
&= \frac{z_{1} - \frac{v}{c}r}{r}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}})\right) \\
&= \left(\cos(\gamma_{\boldsymbol{z}}) - \frac{v}{c} \right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right). \end{split}$$

Henceforth,

$$\begin{split} 1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}') + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) &= 1 + \frac{v}{c}\left(\cos(\gamma_{\boldsymbol{z}}) - \frac{v}{c} \right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \\
&= 1 + \frac{v}{c}\left(\cos(\gamma_{\boldsymbol{z}}) + \frac{v}{c}\cos^{2}(\gamma_{\boldsymbol{z}}) - \frac{v}{c} - \frac{v^{2}}{c^{2}}\cos(\gamma_{\boldsymbol{z}}) \right) + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \\
&\approx 1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \\
&= 1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}). \end{split}$$

Substituting these expressions into $u^{(1)}$, we get:

$$\begin{split} u^{(1)}(t, \boldsymbol{x}') &\approx \int \nu(\boldsymbol{z})P\left(\left(t - \frac{vx_{1}'}{c} - \frac{r}{c} + \frac{vz_{1}}{c^{2}} - \frac{r - z_{1}\frac{v}{c}}{c} \right)\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right) \\
&= \int\nu(\boldsymbol{z})P\left(\left(t - \frac{vx_{1}'}{c} - \frac{r}{c} + \frac{vz_{1}}{c^{2}} - \frac{r}{c} + z_{1}\frac{v}{c^{2}} \right)\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right) \\
&= \int\nu(\boldsymbol{z})P\left(\left(t - \frac{vx_{1}'}{c} - 2\frac{r}{c} + 2\frac{vz_{1}}{c^{2}} \right)\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right). \end{split}$$

Within the $P(\cdot)$ expression, $x_{1}'$ denotes the first coordinate of the antenna when it receives the signal initially, which is the pulse round-trip time (RTT) of the antenna to the target, from the antenna, and reflected back to the antenna. Therefore, we will say $\boldsymbol{x} = \left(0, -L, H \right)$ and $\boldsymbol{x}' = \left(vt, -L, H \right)$, where the orbit of the antenna is estimated as a straight line. Using the Law of Cosines:

$$\begin{split} t &= \frac{r}{c} + \frac{1}{c}\sqrt{r^{2} + \left(vt \right)^{2} - 2rvt\cos(\gamma_{\boldsymbol{z}})} \\
&= \frac{2r\left(c - v\cos(\gamma_{\boldsymbol{z}}) \right)}{c^{2} - v^{2}} \\
&\approx \frac{2r}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right). \end{split}$$

Therefore,

$$\begin{split} vt &\approx v\left(\frac{2r}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \\
&= 2r\frac{v}{c} - 2r\frac{v^{2}}{c^{2}}\cos(\gamma_{\boldsymbol{z}}), \end{split}$$

and

$$x_{1}' \approx 2r\frac{v}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right).$$

Then, within $P(\cdot)$, $-2\frac{r}{c} - \frac{vx_{1}'}{c^{2}}$ becomes

$$\begin{split} -2\frac{r}{c} - \frac{vx_{1}'}{c^{2}} &= -2\frac{r}{c} - \frac{2r}{c}\frac{v^{2}}{c^{2}}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \\
&= -2\frac{r}{c} - 2\frac{r}{c}\frac{v^{2}}{c^{2}} + 2r\frac{v^{3}}{c^{4}}\cos(\gamma_{\boldsymbol{z}}) \\
&\approx -2\frac{r}{c}\left(1 + \frac{v^{2}}{c^{2}} \right) \\
&\approx -2\frac{r}{c}. \end{split}$$

Now, substituting this expression into $u^{(1)}$ we get:

$$\begin{split} u^{(1)}(t, \boldsymbol{x}') &\approx \int\nu(\boldsymbol{z})P\left(\left(t - 2\frac{r}{c} + 2\frac{vz_{1}}{c^{2}} \right)\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right)d\boldsymbol{z} \\
&= \int\nu(\boldsymbol{z})P\left(\left(t - 2\frac{r}{c}\left(1 - \frac{vz_{1}}{cr} \right)\right)\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right)d\boldsymbol{z} \\
&= \int\nu(\boldsymbol{z})P\left(\left(t - 2\frac{r}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right)\right)\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right)d\boldsymbol{z} \\
&\approx \int\nu(\boldsymbol{z})P\left(t \left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}})\right) - 2\frac{r}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right)\right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right)d\boldsymbol{z}. \end{split}$$

This provides a mathematical model for the SAR raw data while taking into account the physical linear Doppler effect in fast time. We also recall that $r$ is the distance between the emitting location of the antenna $\boldsymbol{x}$ and the target $\boldsymbol{z}$: $r = \left|\boldsymbol{z} - \boldsymbol{x}\right| = R_{\boldsymbol{z}}.$ Where $2\frac{r}{c}$ is the retarded moment of time due to travel delay.

Next, we need to apply the matched filter to obtain the imaging kernel in order to define our image. The matched filter portion will have an overbar over it in order to maximize the filter response of $\nu(\boldsymbol{z})$,

$$\begin{split} I_{\boldsymbol{x}}(\boldsymbol{y}) &= \int d\boldsymbol{z}\nu(\boldsymbol{z})\chi_{\Theta}\left(\frac{z_{1} - x_{1}}{R} \right) \cdot \int_{\chi}dt \overline{P\left(t\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}})\right) - 2\frac{R_{\boldsymbol{y}}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}) \right) \right)} \\
&\cdot P\left(t\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) - 2\frac{R_{\boldsymbol{z}}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right), \end{split}$$

where the interior integral is the Point Spread Function (PSF) $W_{\boldsymbol{x}}(\boldsymbol{y}, \boldsymbol{z}).$ The $x_{1}$ argument in $\chi_{\Theta}(\cdot)$ denotes where the radiation signal was emitted from on the antenna, where we will now be looking at multiple emitting locations $\boldsymbol{x}^{n} = (x_{1}^{n}, -L, H).$ The formula for the single-look angles is given by:

$$\begin{split} I_{\boldsymbol{x}^{n}}(\boldsymbol{y}) &= \int d\boldsymbol{z}\nu(\boldsymbol{z})\chi_{\Theta}\left(\frac{z_{1} - x_{1}^{n}}{R} \right) \cdot \int_{\chi}dt \overline{P\left(t\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n})\right) - 2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) \right)} \\
&\cdot P\left(t\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) - 2\frac{R_{\boldsymbol{z}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \right). \end{split}$$

Where,

$$R_{\boldsymbol{y}}^{n} = \left|\boldsymbol{y} - \boldsymbol{x}^{n} \right|, \quad \cos(\gamma_{\boldsymbol{y}}^{n}) = \frac{y_{1} - x_{1}^{n}}{R_{\boldsymbol{y}}^{n}},$$

and

$$R_{\boldsymbol{z}}^{n} = \left|\boldsymbol{z} - \boldsymbol{x}^{n} \right|, \quad \cos(\gamma_{\boldsymbol{z}}^{n}) = \frac{z_{1} - x_{1}^{n}}{R_{\boldsymbol{z}}^{n}}.$$

Where $\boldsymbol{x}^{n}$ is the sequence of emitting locations of the antenna on the orbit that form the synthetic aperture for a given image point $\boldsymbol{y}.$ Then, defining the overall image using the individual contributions of each $I_{\boldsymbol{x}^{n}}(\boldsymbol{y})$ where there is overlap between the image, $y_{1}$, and target location, $z_{1}$, with reference to the emission locations, $x_{1}^{n}$, we obtain:

$$\begin{split} I(\boldsymbol{y}) &= \sum_{n}\chi_{L_{SA}}\left(y_{1} - x_{1}^{n} \right)I_{\boldsymbol{x}^{n}}(\boldsymbol{y}) \\
&= \int\left[\sum_{n}\chi_{L_{SA}}\left(y_{1} - x_{1}^{n} \right)\chi_{L_{SA}}\left(z_{1} - x_{1}^{n} \right)W_{\boldsymbol{x}^{n}}(\boldsymbol{y}, \boldsymbol{z}) \right]\nu(\boldsymbol{z})d\boldsymbol{z} \\
&= \int W(\boldsymbol{y}, \boldsymbol{z})\nu(\boldsymbol{z})d\boldsymbol{z}. \end{split}$$

Where $W(\boldsymbol{y}, \boldsymbol{z})$ is the imaging kernel, which is a coherent sum of the PSFs, $W_{\boldsymbol{x}^{n}}(\boldsymbol{y}, \boldsymbol{z})$ given by the $dt$ integral, previously. Now, we introduce new notation to help with compactness of the formula:

$$\kappa_{\boldsymbol{y}}^{n} = 1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \quad \textrm{and} \quad \kappa_{\boldsymbol{z}}^{n} = 1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}).$$

Substituting $\kappa_{\boldsymbol{y}}^{n}$ into the $\boldsymbol{y}$ term in $P(\cdot)$,

$$t\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) - 2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) = \left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)\left(t - 2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)^{-1} \right).$$

Since, we know $\gamma_{\boldsymbol{y}}^{n}$ & $\gamma_{\boldsymbol{z}}^{n}$ needs to be as close to $\frac{\pi}{2}$ in order to maximize the response of the kernel, we know that $\cos(\gamma_{\boldsymbol{y}}^{n})$ will be very small and the $cos(\gamma_{\boldsymbol{y}}^{n})$ term also has a factor of $\frac{v}{c}$, where $c \gg v$. Therefore, $2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \ll 1$, and using the property $\frac{1}{1 + x} \simeq 1 - x$ when $x \ll 1$. The multiplication on the right-hand side of the second parenthesis term will become:

$$\begin{split} -2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)^{-1} &\approx -2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)\left(1 - 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) \\
&= -2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 - 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) - 2\frac{v^{2}}{c^{2}}\cos^{2}(\gamma_{\boldsymbol{y}}^{n}) \right) \\
&\approx -2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right), \end{split}$$

where we dropped the $\frac{v^{2}}{c^{2}}$ term. Continuing the substitution into the $\boldsymbol{y}$ term in $P(\cdot)$,

$$\begin{split} t\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) - 2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) &\approx \left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)\left(t - 2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)\right) \\
&= \kappa_{\boldsymbol{y}}^{n}\left(t - t_{\boldsymbol{y}}^{n} \right), \end{split}$$

where

$$t_{\boldsymbol{y}}^{n} = 2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) \quad \textrm{and} \quad t_{\boldsymbol{z}}^{n} = 2\frac{R_{\boldsymbol{z}}^{n}}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right).$$

Where the interrogation of the $\boldsymbol{z}$ term in $P(\cdot)$ shows a similar result. Therefore,

$$I_{\boldsymbol{x}^{n}}(\boldsymbol{y}) = \int d\boldsymbol{z}\nu(\boldsymbol{z})\chi_{\Theta}\left(\frac{z_{1} - x_{1}^{n}}{R} \right) \int_{\chi}dt \overline{P\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right)}P\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right),$$

where

$$W_{\boldsymbol{x}^{n}}(\boldsymbol{y}, \boldsymbol{z}) = \int_{\chi}dt \overline{P\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right)}P\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right).$$

Using the following definition, derived in a previous homework,

$$\begin{split} P(t) &= A(t)e^{-i\omega_{o}t} \\
&= \chi_{\tau}(t)e^{-i\alpha t^{2}}e^{-i\omega_{o}t}, \end{split}$$

we can define $W_{\boldsymbol{x}^{n}}(\boldsymbol{y}, \boldsymbol{z})$ as:

$$\begin{split} W_{\boldsymbol{x}^{n}}(\boldsymbol{y}, \boldsymbol{z}) &= \int_{\chi}dt \overline{A\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right)}A\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right)e^{i\omega_{0}\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n}}e^{-i\omega_{0}\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n}} \\
&= \int_{\chi}dt \chi_{\tau}\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right)e^{i\alpha\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right)^{2}} \chi_{\tau}\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right)e^{-i\alpha\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right)^{2}} e^{i\omega_{0}\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n}}e^{-i\omega_{0}\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n}}. \end{split}$$

Finally, defining the entire imaging kernel, $W(\boldsymbol{y}, \boldsymbol{z})$:

$$\begin{split} W(\boldsymbol{y}, \boldsymbol{z}) &= \sum_{n}\chi_{L_{SA}}(y_{1} - x_{1}^{n})\chi_{L_{SA}}(z_{1} - x_{1}^{n}) \cdot W_{\boldsymbol{x}^{n}}(\boldsymbol{y}, \boldsymbol{z}) \\
&= \sum_{n}\chi_{L_{SA}}(y_{1} - x_{1}^{n})\chi_{L_{SA}}(z_{1} - x_{1}^{n}) \\
&\cdot \int_{\chi}dt \chi_{\tau}\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right)e^{i\alpha\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right)^{2}} \chi_{\tau}\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right)e^{-i\alpha\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right)^{2}} e^{i\omega_{0}\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n}}e^{-i\omega_{0}\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n}}. \end{split}$$

Notice the indicator functions, $\chi_{\tau}$, have different arguments in comparison to the Generalized Ambiguity Function (GAF) derived for the non-doppler corrected GAF. This indicates to us the stretching of time that occurs during the RTT of the emitted pulse from the antenna and the reception of the scattered field at the moving antenna. Thus, these intervals that define non-trivial contributions to the imaging kernel will have different lengths between the $\boldsymbol{y}$ and $\boldsymbol{z}$ terms. We define these indicator functions below:

$$\chi_{\tau}\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right) = \begin{cases} 1, & \textrm{if } t \in \left[t_{\boldsymbol{y}}^{n} - \frac{\tau}{2\kappa_{\boldsymbol{y}}^{n}}, t_{\boldsymbol{y}}^{n} + \frac{\tau}{2\kappa_{\boldsymbol{y}}^{n}} \right] \\ 0, & \textrm{otherwise} \end{cases}$$

$$\chi_{\tau}\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right) = \begin{cases} 1, & \textrm{if } t \in \left[t_{\boldsymbol{z}}^{n} - \frac{\tau}{2\kappa_{\boldsymbol{z}}^{n}}, t_{\boldsymbol{z}}^{n} + \frac{\tau}{2\kappa_{\boldsymbol{z}}^{n}} \right] \\ 0, & \textrm{otherwise}. \end{cases}$$

The integration is performed over the intersection of these two intervals, i.e., over $\chi_{\tau}\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right) \cdot \chi_{\tau}\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right)$. Which would be either

$$\left[t_{\boldsymbol{y}}^{n} - \frac{\tau}{2\kappa_{\boldsymbol{y}}^{n}}, t_{\boldsymbol{z}}^{n} + \frac{\tau}{2\kappa_{\boldsymbol{z}}^{n}} \right] \approx \left[t_{\boldsymbol{y}}^{n} - \frac{\tau}{2}\left(1 - 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right), t_{\boldsymbol{z}}^{n} + \frac{\tau}{2}\left(1 - 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \right]$$

or

$$\left[t_{\boldsymbol{z}}^{n} - \frac{\tau}{2\kappa_{\boldsymbol{z}}^{n}}, t_{\boldsymbol{y}}^{n} + \frac{\tau}{2\kappa_{\boldsymbol{y}}^{n}} \right] \approx \left[t_{\boldsymbol{z}}^{n} - \frac{\tau}{2}\left(1 - 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right), t_{\boldsymbol{y}}^{n} + \frac{\tau}{2}\left(1 - 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) \right].$$

The analysis is the same for both cases, so we shall continue with the first interval definition, corresponding to $t_{\boldsymbol{z}}^{n} \lt t_{\boldsymbol{y}}^{n}$. We also introduce the following notation:

$$T^{n} = \frac{t_{\boldsymbol{y}}^{n} - t_{\boldsymbol{z}}^{n}}{2}.$$

Substituting these values and redefining the indicator functions as above, we get the following expression for $W(\boldsymbol{y}, \boldsymbol{z})$:

$$W(\boldsymbol{y}, \boldsymbol{z}) = \sum_{n}\chi_{L_{SA}}(y_{1} - x_{1}^{n})\chi_{L_{SA}}(z_{1} - x_{1}^{n})e^{-i\omega_{0}T^{n}\left(\kappa_{\boldsymbol{y}}^{n} + \kappa_{\boldsymbol{z}}^{n} \right)}\cdot \tau^{n}sinc\left(\tau^{n}\left(\alpha T^{n}\left(\left(\kappa_{\boldsymbol{y}}^{n} \right)^{2} + \left(\kappa_{\boldsymbol{z}}^{n} \right)^{2} \right) - \omega_{0}\frac{\left(\kappa_{\boldsymbol{y}}^{c} - \kappa_{\boldsymbol{z}}^{n} \right)}{2} \right) \right).$$

Since we are interested in the non-factorized & factorized versions of the non-doppler corrected GAF, but the scattered field is still affected by the antenna moving, and thus utilizes the Doppler effect, we know that we need to set the velocity in the $\kappa_{\boldsymbol{y}}^{n}$ to be 0. Using that idea,

$$\begin{split} \kappa_{\boldsymbol{y}}^{n}\bigg\rvert_{v=0} &= 1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \\
&= 1 \end{split}.$$

Thus we obtain our non factorized form of the GAF that has been doppler corrected only with regards to the target, where $W(\boldsymbol{y}, \boldsymbol{z}) = W$ to keep consistent with the equation we are originally solving for and obtain,

$$W = \sum_{n}\chi_{L_{SA}}(y_{1} - x_{1}^{n})\chi_{L_{SA}}(z_{1} - x_{1}^{n})e^{-i\omega_{0}T^{n}\left(1 + \kappa_{\boldsymbol{z}}^{n} \right)}\cdot \tau^{n}sinc\left(\tau^{n}\left(\alpha T^{n}\left(1 + \left(\kappa_{\boldsymbol{z}}^{n} \right)^{2} \right) - \omega_{0}\frac{\left(1 - \kappa_{\boldsymbol{z}}^{n} \right)}{2} \right) \right).$$

To obtain the factorized form of the GAF, we need to acknowledge the fact that the envelope function is varying much slower than the carrier frequency over the summation interval, therefore we choose to take only the center value of the envelope portion, $W_{R}$, where we denote $n = c$:

$$\begin{split} W_{\sum} &= \sum_{n}\chi_{L_{SA}}(y_{1} - x_{1}^{n})\chi_{L_{SA}}(z_{1} - x_{1}^{n})e^{-i\omega_{0}T^{n}\left(1 + \kappa_{\boldsymbol{z}}^{n} \right)} \\
\\
W_{R} &= \tau^{c}sinc\left(\tau^{c}\left(\alpha T^{c}\left(1 + \left(\kappa_{\boldsymbol{z}}^{c} \right)^{2} \right) - \omega_{0}\frac{\left(1 - \kappa_{\boldsymbol{z}}^{c} \right)}{2} \right) \right) \\
\\
W_{(R, \sum)} &= \tau^{c}sinc\left(\tau^{c}\left(\alpha T^{c}\left(1 + \left(\kappa_{\boldsymbol{z}}^{c} \right)^{2} \right) - \omega_{0}\frac{\left(1 - \kappa_{\boldsymbol{z}}^{c} \right)}{2} \right) \right) \cdot \sum_{n}\chi_{L_{SA}}(y_{1} - x_{1}^{n})\chi_{L_{SA}}(z_{1} - x_{1}^{n})e^{-i\omega_{0}T^{n}\left(1 + \kappa_{\boldsymbol{z}}^{n} \right)}. \end{split}$$

Now, we will compute the factorization error, as in [1, (Section 6.6.4)]:

$$\begin{split} W - W_{(R, \sum)} &= \sum_{n}\chi_{L_{SA}}(y_{1} - x_{1}^{n})\chi_{L_{SA}}(z_{1} - x_{1}^{n})e^{-i\omega_{0}T^{n}\left(1 + \kappa_{\boldsymbol{z}}^{n} \right)} \\
& \cdot \bigg[\tau^{n}sinc\left(\tau^{n}\alpha \left(T^{n}\left(1 + \left(\kappa_{\boldsymbol{z}}^{n} \right)^{2} \right) - \omega_{0}\frac{\left(1 - \kappa_{\boldsymbol{z}}^{n} \right)}{2\alpha} \right) \right) \\
& - \tau^{c}sinc\left(\tau^{c}\alpha \left(T^{c}\left(1 + \left(\kappa_{\boldsymbol{z}}^{c} \right)^{2} \right) - \omega_{0}\frac{\left(1 - \kappa_{\boldsymbol{z}}^{c} \right)}{2\alpha} \right) \right)\bigg]. \end{split}$$

We begin solving this by solving for some smaller components,

$$\begin{split} 1 + \left(\kappa_{\boldsymbol{z}}^{n} \right)^{2} &= 1 + \left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right)^{2} \\
&= 1 + 1 + 4\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) + 4\frac{v^{2}}{c^{2}}\cos(\gamma_{\boldsymbol{z}}^{n}) \\
&= 2 + 4\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) + \mathcal{O}\left(\frac{v^{2}}{c^{2}} \right) \end{split}$$

and

$$\begin{split} \frac{1 - \kappa_{\boldsymbol{z}}^{n}}{2} &= \frac{1 - 1 - 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n})}{2} \\
&= -\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \\
&\approx -\frac{v}{c}\frac{z_{1} - x_{1}^{n}}{R}, \quad \textrm{where} \quad \cos(\gamma_{\boldsymbol{z}}^{n}) \approx \frac{z_{1} - x_{1}^{n}}{R}. \end{split}$$

Then, dropping the $\mathcal{O}\left(\frac{v^{2}}{c^{2}} \right)$ terms and remembering the following definitions,

$$T^{n} = \frac{t_{\boldsymbol{y}}^{n} - t_{\boldsymbol{z}}^{n}}{2},$$

$$\begin{split} t_{\boldsymbol{y}}^{n}\bigg\rvert_{v=0} &= \frac{2R_{\boldsymbol{y}}^{n}}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right), \quad t_{\boldsymbol{z}}^{n} = \frac{2R_{\boldsymbol{z}}^{n}}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \\
&= \frac{2R_{\boldsymbol{y}}^{n}}{c} \end{split}$$

$$\begin{split} T^{n} &= \frac{\frac{2R_{\boldsymbol{y}}^{n}}{c} - \frac{2R_{\boldsymbol{z}}^{n}}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right)}{2} \\
&= \frac{2}{c}\frac{R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right)}{2} \\
&= \frac{1}{c}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \right) \end{split}$$

Therefore, substituting in our expressions above, dropping the $\mathcal{O}(\frac{v^{2}}{c^{2}})$ terms, and solving for the following:

$$\begin{split} T^{n}\left(1 + \left(\kappa_{\boldsymbol{z}}^{n} \right)^{2} \right) - \omega_{0}\frac{\left(1 - \kappa_{\boldsymbol{z}}^{n} \right)}{2\alpha} &= \frac{1}{c}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \right)\left(1 + \left(\kappa_{\boldsymbol{z}}^{n} \right)^{2} \right) - \omega_{0}\frac{1 - \kappa_{\boldsymbol{z}}^{n}}{2\alpha} \\
&\approx \frac{1}{c}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \right)\left(2 + 4\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) + \frac{\omega_{0}}{\alpha}\frac{v}{c}\frac{z_{1} - x_{1}^{n}}{R} \\
&= \frac{2}{c}\left(R_{\boldsymbol{y}}^{n} + 2R_{\boldsymbol{y}}^{n}\cos(\gamma_{\boldsymbol{z}}^{n}) - R_{\boldsymbol{z}}^{n}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) - 2R_{\boldsymbol{z}}^{n}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right)\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \\
&+ \frac{\omega_{0}}{\alpha}\frac{v}{c}\frac{z_{1} - x_{1}^{n}}{R} \\
&= \frac{2}{c}\left(R_{\boldsymbol{y}}^{n} + 2R_{\boldsymbol{y}}^{n}\cos(\gamma_{\boldsymbol{z}}^{n}) - R_{\boldsymbol{z}}^{n} + R_{\boldsymbol{z}}^{n}\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) - 2R_{\boldsymbol{z}}^{n}\left(\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) - \frac{v^{2}}{c^{2}}\cos^{2}(\gamma_{\boldsymbol{z}}^{n}) \right) \right) \\
&+ \frac{\omega_{0}}{\alpha}\frac{v}{c}\frac{z_{1} - x_{1}^{n}}{R} \\
&= \frac{2}{c}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n} + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n})\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n} \right) + R_{\boldsymbol{z}}^{n}\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) + \frac{\omega_{0}}{\alpha}\frac{v}{c}\frac{z_{1} - x_{1}^{n}}{R} \\
&\approx \frac{2}{c}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n} + 2\frac{v}{c}\frac{z_{1} - x_{1}^{n}}{R}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n} \right) + R_{\boldsymbol{z}}^{n}\frac{v}{c}\frac{z_{1} - x_{1}^{n}}{R_{\boldsymbol{z}}^{n}} + \frac{\omega_{0}v}{2\alpha}\frac{z_{1} - x_{1}^{n}}{R} \right) \\
&= \frac{2}{c}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n} + 2\frac{v}{c}\frac{z_{1} - x_{1}^{n}}{R}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n} \right) + \frac{v}{c}\left(z_{1} - x_{1}^{n} \right)\left(\frac{R_{\boldsymbol{z}}^{n}}{R_{\boldsymbol{z}}^{n}} + \frac{\omega_{0}c}{2R\alpha} \right) \right) \\
&= \frac{2}{c}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n} + 2\frac{v}{c}\frac{z_{1} - x_{1}^{n}}{R}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n} \right) + \frac{v}{c}\left(z_{1} - x_{1}^{n} \right)\left(1 + \frac{\omega_{0}c}{2R\alpha} \right) \right). \end{split}$$

Since $||R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n}|| \ll R$, the term $2\frac{v}{c}\frac{z_{1} - x_{1}^{n}}{R}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n} \right) \sim 0$ and we can rewrite our above expression as:

$$T^{n}\left(1 + \left(\kappa_{\boldsymbol{z}}^{n} \right)^{2} \right) - \omega_{0}\frac{\left(1 - \kappa_{\boldsymbol{z}}^{n} \right)}{2\alpha} \approx \frac{2}{c}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n} + \frac{v}{c}\left(z_{1} - x_{1}^{n} \right)\left(1 + \frac{\omega_{0}c}{2R\alpha} \right) \right).$$

Using the derivation that $R_{Dop} \approx -2d - 2\frac{l\sin(\theta)}{R}\tilde{x}$ and $\left(z_{1} - x_{1}^{n} \right)\left(1 + \frac{\omega_{0}c}{2R\alpha} \right) = R_{Dop}$,

$$\begin{split} T^{n}\left(1 + \left(\kappa_{\boldsymbol{z}}^{n} \right)^{2} \right) - \omega_{0}\frac{\left(1 - \kappa_{\boldsymbol{z}}^{n} \right)}{2\alpha} &\approx \frac{2}{c}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n} - 2\frac{v}{c}\frac{l\sin(\theta)}{R}\tilde{x} \right) \\
&\approx \frac{2}{c}\left(\frac{Ll}{R} - \frac{v}{c}\frac{y_{1} - z_{1}}{2} \right) - \frac{v}{c}\left(y_{1} - z_{1} \right)\frac{\omega_{0}}{2\alpha R} - \frac{2}{c}\left(\frac{y_{1} - z_{1}}{R} + \frac{v}{c}\left(1 + \frac{\omega_{0}}{\alpha R}\frac{c}{2} \right) \right)\tilde{x}. \end{split}$$

Where, $L$ is the projected vector from the antenna to the ground in the range direction, $l = y_{2} - z_{1}$ & $\sin(\theta) = \frac{L}{R}$.

We then introduce the definition $T'^{n} = T'^{c} - \mathcal{T}'^{\tilde{n}}$, where

$$\begin{split} T'^{c} &= T^{c}\left(1 + \left(\kappa_{\boldsymbol{z}}^{c} \right)^{2} \right) - \omega_{0}\frac{1 - \kappa_{\boldsymbol{z}}^{c}}{2\alpha} \\
&= \frac{2}{c}\left(\frac{Ll}{R} - \frac{v}{c}\frac{y_{1} - z_{1}}{2} \right) - \frac{v}{c}\left(y_{1} - z_{1} \right)\frac{\omega_{0}}{2\alpha R} \\
\\
\mathcal{T}'^{1} &= \frac{2}{c}\left(\frac{y_{1} - z_{1}}{R} + \frac{v}{c}\left(1 + \frac{\omega_{0}}{\alpha R}\frac{c}{2} \right) \right)\Delta x_{1} \\
\\
\mathcal{T}'^{\tilde{n}} &= \frac{2}{c}\left(\frac{y_{1} - z_{1}}{R} + \frac{v}{c}\left(1 + \frac{\omega_{0}}{\alpha R}\frac{c}{2} \right) \right)\tilde{x} \\
&= \tilde{n}\mathcal{T}'^{1}. \end{split}$$

Using the values in Table 1.1, we see that typically $|T'^{c}| \ll \tau$, where $\tau = \frac{B}{2\alpha} \sim 5\cdot10E-5s$. Additionally, we see that this same equality holds true for $|\mathcal{T}'^{1}| \ll \tau$. This allows us to use the first order Taylor expansion in our $W-W_{(R, \sum)}$ factorization error expression, which we can now rewrite as:

$$\begin{split} W - W_{(R, \sum)} &= \sum_{n}\chi_{L_{SA}}\left(y_{1} - x_{1}^{n} \right)\chi_{L_{SA}}\left(z_{1} - x_{1}^{n} \right)e^{-i\omega_{0}T^{n}\left(1 + \kappa_{\boldsymbol{z}}^{n} \right)} \\
&\cdot \left[\tau^{n}sinc\left(\alpha\tau^{n}T'^{n} \right) - \tau^{c}sinc\left(\alpha\tau^{c}T'^{c} \right) \right]. \end{split}$$

Using the first order Taylor expansion of our $sinc$ function, we get:

$$\begin{split} sinc(\alpha\tau^{\tilde{n}}T'^{\tilde{n}}) &\approx sinc(\alpha\tau^{c}T'^{c}) + \alpha(\tau^{\tilde{n}}T'^{\tilde{n}})sinc'(\alpha\tau^{c}T'^{c}) \\
&\approx sinc(\alpha\tau^{c}) - \alpha\tau\mathcal{T}'^{\tilde{n}}sinc'(\alpha\tau^{c}T'^{c}). \end{split}$$

Then, the $sinc$ terms in $W - W_{(R, \sum)}$ become:

$$\tau^{\tilde{n}}sinc(\alpha\tau^{\tilde{n}}T'^{\tilde{n}}) - \tau^{c}sinc(\alpha\tau^{c}T'^{c}) \approx 2\left(\left|T^{c} \right| - \left|T^{\tilde{n}} \right| \right)sinc\left(\alpha\tau^{c}T'^{c} \right) - \alpha\tau^{2}\mathcal{T}'^{\tilde{n}}sinc'\left(\alpha\tau^{c}T'^{c} \right).$$

We will also define the following values:

$$\begin{split} S' &= -\alpha\tau^{2}sinc'(\alpha\tau^{c}T'^{c}) \\
\\
\varphi' &= 2k_{0}\left(\frac{y_{1} - z_{1}}{R} + \frac{v}{c} \right)\frac{L_{SA}}{N} \\
\\
\Phi_{0} &= -2k_{0}\frac{Ll}{R}. \end{split}$$

Then, we can rewrite the factorization error as:

$$\begin{split} W - W_{(R, \sum)} &\approx e^{i\Phi_{0}}S'\sum_{\tilde{n} = -\frac{\tilde{N}}{2}}^{\frac{\tilde{N}}{2}}e^{-i\omega_{0}T^{n}\left(1 + \kappa_{\boldsymbol{z}}^{n} \right)}\mathcal{T}'^{\tilde{n}} \\
&= S'\mathcal{T}'^{1}e^{i\Phi_{0}}\sum_{\tilde{n} = -\frac{\tilde{N}}{2}}^{\frac{\tilde{N}}{2}}\tilde{n}e^{i\tilde{n}\varphi'}. \end{split}$$

Where the summation is a geometric series that converges to a sinc function:

$$\begin{split} \sum_{\tilde{n} = -\frac{\tilde{N}}{2}}^{\frac{\tilde{N}}{2}}\tilde{n}e^{i\tilde{n}\varphi'} &= \frac{1}{i}\frac{\partial}{\partial\varphi'}\sum_{\tilde{n} = -\frac{\tilde{N}}{2}}^{\frac{\tilde{N}}{2}}e^{i\tilde{n}\varphi'} \\
&\approx \frac{1}{i}\frac{\partial}{\partial\varphi'}\left(\tilde{N}sinc\left(\frac{\tilde{N}\varphi'}{2} \right) \right) \\
&= \frac{\tilde{N}^{2}}{2i}sinc'\left(\frac{\tilde{N}\varphi'}{2} \right). \end{split}$$

Thus, the factorization error becomes:

$$\begin{split} W - W_{(R, \sum)} &\approx S'T'^{1}\frac{N^{2}}{2i}e^{i\Phi_{0}}sinc'\left(\frac{N\varphi'}{2} \right) \\
&= \frac{NS'}{2i\omega_{0}}e^{i\Phi_{0}}\frac{N\varphi'}{2}sinc'\left(\frac{N\varphi'}{2} \right). \end{split}$$

Using the following definition $y_{1} - z_{1} = \frac{\pi R}{k_{0}L_{SA}} = \frac{\pi Rc}{\omega_{0}L_{SA}} = \Delta_{A}$ to solve for $\frac{N\varphi'}{2}$:

$$\begin{split} \frac{N\varphi'}{2} &= \frac{N}{2}\left(2k_{0}\left(\frac{y_{1} - z_{1}}{R} + \frac{v}{c} \right)\frac{L_{SA}}{N} \right) \\
&= L_{SA}k_{0}\left(\frac{y_{1} - z_{1}}{R} + \frac{v}{c} \right) \\
&= \frac{L_{SA}k_{0}}{\pi R}\left(\left(y_{1} - z_{1} \right)\pi + \pi R\frac{v}{c} \right) \\
&= \frac{\pi}{\Delta_{A}}\left(\left(y_{1} - z_{1} \right) + R\frac{v}{c} \right). \end{split}$$

Then, we can write the factorization error as:

$$\frac{\max\left|W - W_{(R, \sum)} \right|}{\max\left|W_{(R, \sum)} \right|} = \frac{1}{N\tau}\frac{N\left|S' \right|}{\omega_{0}}\left|\frac{N\omega_{0}\mathcal{T}'^{1}}{2} \right|\left|sinc'\left(\frac{N\varphi'}{2} \right) \right|.$$

Taking into account that

$$\max\limits_{x \in \left(-\infty, \infty \right)}\left|sinc'(x) \right| \lt \frac{1}{2}$$

and substituting in our previous definitions, we get:

$$\frac{\max\left|W - W_{(R, \sum)} \right|}{\max\left|W_{(R, \sum)} \right|} \lesssim \frac{B}{8\omega_{0}}\frac{\pi}{\Delta_{A}}\left|\left(y_{1} - z_{1} \right) + \frac{v}{c}R + \frac{\omega_{0}c}{2\alpha R}\frac{v}{c}R \right|.$$

Now, we will look at some values from [1, (Table 1.1)], [1, (Table 1.2)], & [1, (Table 1.3)] that are typical of the values in the approximate upper bound of the factorization error.

$$\begin{split} \frac{\omega_{0}}{2\pi} &\sim 300MHz = 3\cdot10^{8}Hz \approx 3.33\cdot10^{-9}s \\
\omega_{0} &\sim 4.77\cdot10^{7}Hz \approx 2.09\cdot10^{-8}s \end{split}$$

$$\begin{split} v &\sim 7.6km/s = 7.6\cdot10^{3}m/s \\
c &\sim 3\cdot10^{10}cm/s = 3\cdot10^{8}m/s \\
\frac{v}{c} &\sim 2.53\cdot10^{-5} \end{split}$$

$$R \sim 1000km = 1\cdot10^{6}m$$

$$\left|l \right| = \left|y_{2} - z_{2} \right| \sim \left|y_{1} - z_{1} \right| \sim 10m$$

$$\begin{split} \frac{B}{\omega_{0}} &\sim 3\cdot10^{-2} \\
B &\sim 1.43\cdot10^{6}Hz \end{split}$$

$$\begin{split} \tau &\sim 5\cdot10^{-4}s \\
\tau &= \frac{B}{2\alpha} \\
\alpha &= \frac{B}{2\tau} \sim 6.98\cdot10^{-4}s \end{split}$$

$$\Delta_{A} \sim 10m$$

Plugging these values into our above expression and canceling out the $R$'s in the last term inside $\left|\cdot \right|$, we get:

$$\begin{split} \frac{\max\left|W - W_{(R, \sum)} \right|}{\max\left|W_{(R, \sum)} \right|} &\lesssim \frac{B}{8\omega_{0}}\frac{\pi}{\Delta_{A}}\left|10m + \left(2.53\cdot10^{-5} \right)\left(1\cdot10^{6}m \right) + \frac{2.09\cdot10^{-8}s\left(3\cdot10^{8}m/s \right)}{2\left(6.98\cdot10^{-4}s \right)}2.53\cdot10^{-5}s \right| \\
&\lesssim \frac{\pi}{8}\frac{B}{\omega_{0}}\left(\frac{1}{\Delta_{A}}\left|10m + 25.3m + 0.1136m \right| \right) \\
&\lesssim \frac{\pi}{8}\frac{B}{\omega_{0}}\left|\frac{10m}{\Delta_{A}} + \frac{25.3m}{\Delta_{A}} + \frac{0.1136m}{\Delta_{A}} \right| \\
&\lesssim \frac{\pi}{8}\frac{B}{\omega_{0}}\left|\mathcal{O}(1) + \mathcal{O}(1) + \mathcal{O}\left(\ll 1 \right) \right|. \end{split}$$

Therefore, we can write the factorization error as:

$$\frac{\max\left|W - W_{(R, \sum)} \right|}{\max\left|W_{(R, \sum)} \right|} \lesssim \frac{\pi}{8}\frac{B}{\omega_{0}}\cdot const.$$

However, this is just for typical values, where $\frac{\left|y_{1} - z_{1} \right|}{\Delta_{A}} \lesssim 1$, the second term $\frac{v}{c}R$ will most likely not change much in practice due to there being a typical antenna moving velocity, $v$, the speed of light being a constant and a typical slant one-way distance from the orbit to the target, $R$. The third term can be more volatile depending on $\omega_{0}$ & $\alpha$, however, since $\omega_{0}$ does end up being a parameter in the azimuthal resolution, so we can see that an increase in $\omega_{0}$, a slower frequency, will begin to drastically increase the factorization error of this term, since when it is divided by $\Delta_{A} = \frac{\pi R}{k_{0}L_{SA}} = \frac{\pi Rc}{\omega_{0}L_{SA}}$, $\omega_{0}$ will become squared and lowering the frequency of the carrier frequency will have a large impact on this third term. Additionally, the envelope frequency or chirp rate, $\alpha$, can have a large impact on this third term of the factorization error, where if it grows too small (a faster frequency), then we begin to see a large growth in this factorization term.

So, the main term to be aware of, that is more volatile than the others and may contribute to a larger factorization error, is the third term $\frac{\omega_{0}c}{2\alpha}\frac{v}{c}\frac{1}{\Delta_{A}}$. And when this does not simply become $\ll 1$, we get a new factorization error that is approximately:

$$\begin{split} \frac{B}{8\omega_{0}}\frac{\pi}{\Delta_{A}}\frac{\omega_{0}c}{2\alpha}\frac{v}{c} &= \frac{\tau}{8}\frac{\omega_{0}L_{SA}}{\pi R}\pi\frac{v}{c} \\
&= \frac{1}{8}\omega_{0}\tau\frac{v}{c}\frac{L_{SA}}{R} \\
&= \frac{1}{4}\tau\Delta\omega_{max}. \end{split}$$

Where,

$$\begin{split} \Delta\omega_{max} &= \omega_{0}\frac{L_{SA}}{2R}\frac{v}{c} = k_{0}\frac{L_{SA}}{2R}v \\
&= \frac{1}{s}\frac{m/s}{m/s}\frac{m}{m} = \frac{1}{s} = Hz \end{split}$$

indicates the maximum absolute variation of the Doppler frequency shift over the length of the synthetic array. Which, if we plug in the typical values we found from Table 1.1 previously, and identify the typical value $L_{SA} \sim 50km = 5\cdot10^{4}m$, we see the typical value for this maximum absolute variation of the Doppler frequency shift as:

$$\begin{split} \Delta\omega_{max} &\approx 4.77\cdot10^{7}Hz\left(\frac{5\cdot10^{4}m}{2\cdot10^{6}m} \right)2.53\cdot10^{-5} \\
&\approx 30.17Hz. \end{split}$$

This maximum absolute variation of the frequency shift has the potential to be a large source of error with regards to inspecting the factorization error and cause image distortions due to not utilizing the non-factorized GAF. In practice, however, we do not utilize this factorization error as the image processing algorithms do not utilize the form $I(\boldsymbol{y}) = \int\nu(\boldsymbol{z})W(\boldsymbol{y}, \boldsymbol{z})d\boldsymbol{z}$ to calculate the overall image $I(\boldsymbol{y})$. Instead, the algorithms utilize $I_{\boldsymbol{x}}(\boldsymbol{y}) = \int_{\chi}\overline{P\left(t - 2\frac{R_{\boldsymbol{y}}}{c} \right)}u^{(1)}(t, \boldsymbol{x})dt$ to first calculate the successive images from the received scattered field at the antenna using the matched filter. This is then followed by summing up all of these independent images over the overlap between the target, $z_{1}$, and image, $y_{1}$ by $I(\boldsymbol{y}) = \sum_{n}\chi_{L_{SA}}(y_{1} - x_{1}^{n})I_{\boldsymbol{x}^{n}}(\boldsymbol{y})$. Therefore, we do not use the GAF in a direct way when processing the images, rather the GAF is a mathematical tool to investigate the impacts of the SAR imagery under different circumstances, as in this case, identifying the error in using a factorized, non-doppler corrected form of the GAF on a scattered field that is affected by the Doppler frequency shift.

A significant assumption that is made in typical SAR imaging is the start-stop approximation, which neglects all of the $\sim \frac{v}{c}$ terms in the matched filter, which makes the chirp rate, envelope frequency, independent of either $\boldsymbol{y}$ or $n$. Therefore, when not taking the Doppler corrections into account, the chirp rate is constant, since if we implemented the corrections, the chirp rate would need to differ per successive image, $I_{\boldsymbol{x}^{n}}(\boldsymbol{y})$ to account for the differences in time between the expected round-trip time and the true round-trip time from a moving source. This enables "range compression" to be performed once over the entire dataset rather than individually per $\boldsymbol{y}$, thus, allowing the computation to be performed by a FFT rather than a convolution in the time space - which reduces the computational complexity.

However, as we saw above, this factorization error can manifest substantial image distortions due to the shift in the argument of the $sinc(\cdot)$ of $W(\boldsymbol{y}, \boldsymbol{z}) \approx \sum_{n}\chi_{L_{SA}}(y_{1} - x_{1}^{n})\chi_{L_{SA}}(z_{1} - x_{1}^{n})e^{-i\omega_{0}T^{n}\left(1 + \kappa_{\boldsymbol{z}}^{n} \right)}\cdot\tau^{n}sinc\left(\tau^{n}\alpha\left(T^{n}\left(1 + \left(\kappa_{\boldsymbol{z}}^{n} \right)^{2} \right) - \omega_{0}\frac{1 - \kappa_{\boldsymbol{z}}^{n}}{2\alpha} \right) \right)$ where the shift is manifested via $1 - \kappa_{\boldsymbol{z}}^{n}$. This means that the shift depends on $n$, as evidenced by $1 - \kappa_{\boldsymbol{z}}^{n}$, and when summing up these $sinc$ terms, we see a "misalignment" between the $I_{\boldsymbol{x}^{n}}(\boldsymbol{y})$ images and retrieve a blurred version of the images to generate the overall image $I(\boldsymbol{y})$. We can see this phenomenon rear its head in the factorization error that we derived above in the term $\frac{1}{\propto} \alpha$.

Additionally, we see an azimuthal shift of magnitude $R\frac{v}{c}$ that is seen in the factorized form of the GAF where $W_{\sum} \approx e^{i\Phi_{0}}Nsinc\left(\frac{\pi}{\Delta_{A}}\left(\left(y_{1} - z_{1} \right) + R\frac{v}{c} \right) \right)$ from an earlier derivation. This shift is unique to the blurring artifact as we can directly see this azimuthal shift embedded in the factorized form of the GAF and we can only observe the blurring origins from the factorization error derivation.
