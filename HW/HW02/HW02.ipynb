## Problem 1.

Show that, the functions $Asinc(Ax)$ converge weakly to $\pi\delta(x)$ as $A\to\infty$. In other words, show that for any smooth compactly supported function $\phi(x)$ (a Schwartz test function), we have:

$$\int_{-\infty}^{\infty}Asinc(Ax)\phi(x)dx \to \pi \phi(0) \quad \textrm{as} \quad A \to \infty.$$

See the argument in [1, Section 3.3].

From [4], we define our Schwartz test function as,

$$\left|\frac{\partial^{k}\phi(x)}{\partial x^{k}}\right| \leq C_{m, k}(1 + |x|)^{-m}$$

for any k and any positive integer m. As a direct consequence of this definition, Schwartz class functions are $C^{\infty}$ functions whose derivatives decay faster than any polynomial.

Evaluating our original integral,

$$\int_{-\infty}^{\infty}Asinc(Ax)\phi(x)dx = A\int_{-\infty}^{\infty}sinc(Ax)\phi(x)dx,$$

where the values inside the integrand in the right hand side equation is the Fourier Transform in x of the product $sin(Ax)\phi(x)$. Now, performing a change of variables where $\omega = Ax \Rightarrow x = \frac{\omega}{A}$ & $d\omega = Adx \Rightarrow dx = \frac{d\omega}{A}$ and the bounds will not change when determining new bounds with these change of variables, because they are $\left(-\infty, \infty\right)$, we get:

$$\begin{split} &= A\int_{-\infty}^{\infty}sinc(\omega)\phi(\frac{\omega}{A})\frac{d\omega}{A} \\
&= \int_{-\infty}^{\infty}sinc(\omega)\phi({\frac{\omega}{A}})d\omega \\
&= \int_{-\infty}^{\infty}sinc(\omega)\phi(0)d\omega, \quad \text{where} \quad \frac{\omega}{A} \to 0 \\
&= \phi(0)\int_{-\infty}^{\infty}sinc(\omega)d\omega. \end{split}$$

Where we are able to say ${\frac{\omega}{A}} \to 0$ due to our original statement of showing weak convergence as $A \to \infty$. Now we need to solve for $\int_{-\infty}^{\infty}sinc(\omega)d\omega$ by,

$$\int_{-\infty}^{\infty}sinc(\omega)d\omega = 2\int_{0}^{\infty}sinc(\omega)d\omega,$$

because $sinc$ is an even function. We then define,

$$\begin{split} G(t) &= \int_{0}^{\infty}sinc(\omega)e^{-t\omega}d\omega \\
&= \int_{0}^{\infty}\frac{\sin(\omega)}{\omega}e^{-t\omega}d\omega, \end{split}$$

where $sinc(\omega) = \frac{\sin(\omega)}{\omega}$ and we wish to evaluate $G(t)$ at $t=0$ as that would be equivalent to our original problem. Now we differentiate $G(t)$ with respect to $t$,

$$\begin{split} \frac{dG(t)}{dt} &= \frac{d}{dt}\int_{0}^{\infty}\frac{\sin(\omega)}{\omega}e^{-t\omega}d\omega \\
&= \int_{0}^{\infty}\frac{d}{dt}e^{-t\omega}\frac{\sin(\omega)}{\omega}d\omega \\
&= \int_{0}^{\infty} \frac{\sin(\omega)}{\omega}\cdot\left(-\omega e^{-t\omega}\right)d\omega \\
&= -\int_{0}^{\infty}\sin(\omega)e^{-t\omega}d\omega. \end{split}$$

We can see the final result above is the negative Laplace transform of $\sin(\omega)$. Using integration by parts, we get:

$$\begin{split} \frac{dG(t)}{dt} &= e^{-t\omega}\cos(\omega)\bigg\rvert_{\omega=0}^{\omega=\infty} + te^{-t\omega}\sin(\omega)\bigg\rvert_{\omega=0}^{\omega=\infty} + \int_{0}^{\infty}t^{2}e^{-t\omega}\sin(\omega)d\omega \\
&= e^{-t\omega}\cos(\omega)\bigg\rvert_{\omega=0}^{\omega=\infty} + te^{t\omega}\sin(\omega)\bigg\rvert_{\omega=0}^{\omega=\infty} - t^{2}\frac{dG(t)}{dt} \\
&= (0 - 1) + (0 - 0) - t^{2}\frac{dG(t)}{dt} \\ 
\frac{dG(t)}{dt} &= -1 - t^{2}\frac{dG(t)}{dt} \\
\frac{dG(t)}{dt} &= -\frac{1}{1 + t^{2}} \end{split}$$

Now we integrate over $\frac{dG(t)}{dt}$ to get back to $G(t)$,

$$\begin{split} G(t) &= \int\frac{dG(t)}{dt}dt \\
&= -\int\frac{1}{1+t^{2}}dt \\
&= -\tan^{-1}(t) + \frac{\pi}{2}. \end{split}$$

At $t = 0$, $\tan^{-1}(t=0) \to 0$, thus, $G(t=0) = \frac{\pi}{2}$. Evaluating our original problem,

$$\begin{split} \int_{-\infty}^{\infty}sinc(\omega)d\omega &= 2\int_{0}^{\infty}sinc(\omega)d\omega \\
&= 2G(0) \\
&= 2\frac{\pi}{2} \\
&= \pi, \end{split}$$

and thus,

$$\begin{split} \int_{-\infty}^{\infty}Asinc(Ax)\phi(x)dx &= \phi(0)\int_{-\infty}^{\infty}sinc(\omega)d\omega \\
&\to \pi \phi(0) \end{split}$$

## Problem 2.

Show that, if the length of the synthetic aperture $L_{SA}$ is chosen equal to the size of the antenna footprint on the ground, see [2, equation (2.29)], then the resulting azimuthal resolution $\Delta_{A}$ depends neither on the carrier frequency $\omega_{o}$ nor the distance to the target $R$. Also demonstrate that, the resolution $\Delta_{A}$ improves as the antenna size $D$ becomes smaller. See the discussion in [3, Section 5.1].

Defining our image as $I(y)$ from applying a matched filter to our original radiation pattern, $P\left(t - \frac{2\left|\underline{y} - \underline{x}\right|}{c} \right)$, we get

$$I(y) = \int \left[\sum_{n} \chi_{L_{SA}}(y_{1} - x_{1}^{n}) \chi_{L_{SA}}(z_{1} - x_{1}^{n}) W_{x^{n}}(\underline{y}, \underline{z}) \right] \nu(\underline{z})d\underline{z}. \quad \textrm{(1).} \quad$$

Where, the $n$ indicates the successive location where the antenna is emitting/receiving radiation according to the Start-Stop approximation, i.e., the slow time, $\nu(\underline{z})$ is the ground reflectivity, $\underline{x}$ is the 3D position of the antenna, $\underline{z}$ is the 3D position of the target & $\underline{y}$ is the 3D position of the image, in cartesian coordinates.

$$\chi_{L_{SA}}(y_{1} - x_{1}^{n}) = \begin{cases} 1, & x_{1}^{n} \in \left[y_{1} - \frac{L_{SA}}{2}, y_{1} + \frac{L_{SA}}{2} \right] \\ 0, & \text{otherwise} \end{cases}$$

$$\chi_{L_{SA}}(z_{1} - x_{1}^{n}) = \begin{cases} 1, & x_{1}^{n} \in \left[z_{1} - \frac{L_{SA}}{2}, z_{1} + \frac{L_{SA}}{2} \right] \\ 0, & \text{otherwise} \end{cases}$$

$$W_{x^{n}}(\underline{y}, \underline{z}) = e^{2i\omega_{o}\left(\frac{R_{\underline{z}}^{n} - R_{\underline{y}}^{n}}{c}\right)}\int_{\chi} \overline{A\left(t - t_{n} - \frac{2R_{\underline{y}}^{n}}{c} \right)} A\left(t - t_{n} - \frac{2R_{\underline{z}}^{n}}{c}dt \right).$$

$$\nu(\underline{z}) = \frac{\omega_{o}^{2}}{16\pi^{2}R^{2}}\frac{n^{2}(\underline{z}) - 1}{c^{2}} \quad \textrm{and} \quad n^{2} = n^{2}(\underline{z}) = \epsilon \mu$$

Where $\epsilon$ is the electric permittivity of the material on the ground & $\mu$ is the magnetic permeability of the material on the ground. And define,

$$R_{y}^{n} = \lVert\underline{y} - \underline{x}^{n}\rVert$$
$$R_{z}^{n} = \lVert\underline{z} - \underline{x}^{n}\rVert.$$

Where we are interested in the terms inside of the brackets in (1)., which is called the Generalized Ambiguity Function (GAF), $W(\underline{y}, \underline{z})$:

$$W(\underline{y}, \underline{z}) = \sum_{n}\chi_{L_{SA}}(y_{1} - x_{1}^{n})\chi_{L_{SA}}(z_{1} - x_{1}^{n})W_{x^{n}}(\underline{y}, \underline{z}).$$

Where we are interested, specifically, in the terms in the direction parallel to the flight track, $W_{\sum}$:

$$W_{\sum} = \sum_{n}\chi_{L_{SA}}(y_{1} - x_{1}^{n})\chi_{L_{SA}}(z_{1} - x_{1}^{n})e^{2i\omega_{o}\left(\frac{R_{\underline{z}^{n}}}{c} - \frac{R_{\underline{y}}^{n}}{c} \right)}.$$

We start by approximating $R_{\underline{z}}^{n}$ & $R_{\underline{y}}^{n}$, using the Taylor Series approximation:

$$\begin{split} R_{\underline{z}}^{n} &= \left(L^{2} + H^{2} + \left(x_{1}^{n} - z_{1} \right)^{2} \right)^{\frac{1}{2}} \\
&= \left(R^{2} + \left(x_{1}^{n} - z_{1} \right)^{2} \right)^{\frac{1}{2}} \\
&\approx R + \frac{1}{2} \frac{\left(x_{1}^{n} - z_{1} \right)^{2}}{R} \end{split}$$

$$\begin{split} R_{\underline{y}}^{n} &= \left(\left(L + l \right)^{2} + H^{2} + \left(x_{1}^{n} - y_{1} \right)^{2} \right)^{\frac{1}{2}}; \quad y_{2} - z_{2} = l \quad \\
&\approx R + \frac{1}{2}\frac{2Ll + l^{2}\cos^{2}(\theta) + \left(x_{1}^{n} - y_{1} \right)^{2}}{R}. \end{split}$$

Where $\theta$ is the elevation angle of the antenna, and we utilized the law of cosines:

$$\begin{split} R_{\underline{z}}^{n} - R_{\underline{y}}^{n} &= \left(R + \frac{1}{2}\frac{\left(x_{1}^{n} - z_{1} \right)^{2}}{R} \right) - \left(R + \frac{2Ll}{2R} + \frac{l^{2}\cos^{2}(\theta)}{2R} + \frac{\left(x_{1}^{n} - y_{1} \right)^{2}}{2R} \right) \\
&= \frac{1}{2}\frac{\left(x_{1}^{n} - z_{1} \right)^{2}}{R} - \frac{1}{2}\frac{\left(x_{1}^{n} - y_{1} \right)^{2}}{R} - \frac{Ll}{R} - \frac{l^{2}\cos^{2}(\theta)}{2R} \\
&= -\frac{Ll}{R} + \frac{z_{1}^{2} - y_{1}^{2} - l^{2}\cos^{2}(\theta)}{2R} + \frac{\left(y_{1} - z_{1} \right)x_{1}^{n}}{R}. \end{split}$$

Now, we define the change in antenna position in the flight track as $\Delta x_{1} = \frac{L_{SA}}{N}$, where $N$ is the total number of pulses in slow time & define $x_{1}^{n} = n\Delta x_{1}.$ We are only interested in the antenna footprint area we get according to the $\chi_{L_{SA}}$ indicator functions that overlap between the image, $y_{1}$, and the target $z_{1},$ seen in the figure below, where $n^{c}$ is the centered slow time index. Thus, we can determine the slow times that contribute to the the summation, that are non-zero, to go from $\left(N_{1}, N_{2} \right):$

$$N_{1} = \left[\frac{max(y_{1}, z_{1})}{\Delta x_{1}} - \frac{L_{SA}}{2\Delta x_{1}} \right], \quad N_{2} = \left[\frac{min(y_{1}, z_{1})}{\Delta x_{1}} + \frac{L_{SA}}{2\Delta x_{1}} \right].$$

$$\frac{y_{1}}{\Delta x_{1}} - \frac{N}{2} \leq n \leq \frac{y_{1}}{\Delta x_{1}} + \frac{N}{2}, \quad \text{where} \quad N = \left[\frac{2\lambda_{o}R}{\Delta x_{1}D} \right]$$



We rewrite $W_{\sum}$ using the above figure to construct a new indicator function that is non-zero within the shaded region, overlap of the antenna footprint between the image and the target, using the new slow time bounds and substituting the wave number, $k_{o} = \frac{\omega_{o}}{c}$ into the exponent:

$$W_{\sum}(\underline{y}, \underline{z}) = \chi_{2L_{SA}}(y_{1} - z_{1})\sum_{n=N_{1}}^{n=N_{2}}e^{2ik_{o}\left(R_{\underline{z}}^{n} - R_{\underline{y}}^{n} \right)}$$

$$\chi_{2L_{SA}}(y_{1} - z_{1}) = \begin{cases} 1, & -L_{SA} \leq y_{1} - z_{1} \leq L_{SA} \\ 0, & \text{otherwise} \end{cases}$$

Now, we redefine the summation term, $W_{\sum}(\underline{y}, \underline{z})$, with our modified slow time with reference to the center value, $n^{c} = \frac{y_{1} + z_{1}}{2\Delta x_{1}},$ and our expression for $R_{\underline{z}}^{n} - R_{\underline{y}}^{n}$:

$$\tilde{n} = n - n^{c}, \quad \tilde{N} = N - \left[\frac{\left|y_{1} - z_{1} \right|}{\Delta x_{1}} \right]$$

$$W_{\sum}(\underline{y}, \underline{z}) = e^{-2ik_{o}\frac{Ll}{R}}e^{-2ik_{o}\frac{l^{2}\cos^{2}(\theta)}{R}} \chi_{2L_{SA}}(y_{1} - z_{1}) \sum_{\tilde{n} = -\frac{\tilde{N}}{2}}^{\tilde{n} = \frac{\tilde{N}}{2}}e^{2ik_{o}\frac{x_{1}^{n}(y_{1} - z_{1})}{R}}e^{ik_{o}\frac{z_{1}^{2} - y_{1}^{2}}{R}}.$$

Where $\tilde{N}$ is the slow times within the overlap in the above figure that are actually used for image reconstruction. Substituting $x_{1}^{n} = \frac{nL_{SA}}{N} = \frac{(\tilde{n} + n^{c})L_{SA}}{N}$ and $n^{c} = \frac{y_{1} + z_{1}}{2\Delta x_{1}}$ we get:

$$W_{\sum}(\underline{y}, \underline{z}) = e^{-2ik_{o}\frac{Ll}{R}}e^{-2ik_{o}\frac{l^{2}\cos^{2}(\theta)}{R}} \chi_{2L_{SA}}(y_{1} - z_{1}) \sum_{\tilde{n} = -\frac{\tilde{N}}{2}}^{\tilde{n} = \frac{\tilde{N}}{2}}e^{2ik_{o}\frac{\left(\tilde{n} + \frac{y_{1} + z_{1}}{2\Delta x_{1}} \right)(y_{1} - z_{1})}{R}}e^{ik_{o}\frac{z_{1}^{2} - y_{1}^{2}}{R}}.$$

Then, solving for the expression in the first exponent in the summation, we see:

$$\begin{split} \left(\tilde{n} + \frac{y_{1} + z_{1}}{2\Delta x_{1}} \right)(y_{1} - z_{1}) &= \tilde{n}y_{1} + y_{1}\frac{y_{1} + z_{1}}{2\Delta x_{1}} - z_{1}\tilde{n} - z_{1}\frac{y_{1} + z_{1}}{2\Delta x_{1}} \\
&= \tilde{n}(y_{1} - z_{1}) + \frac{y_{1}^{2} + y_{1}z_{1}}{2\Delta x_{1}} - \frac{y_{1}z_{1} - z_{1}^{2}}{2\Delta x_{1}} \\
&= \tilde{n}(y_{1} - z_{1}) + \frac{y_{1}^{2} - z_{1}^{2}}{2\Delta x_{1}}. \end{split}$$

Substituting this expression back into the first exponent, we see:

$$\begin{split} e^{2ik_{o}\frac{\left(\tilde{n} + \frac{y_{1} + z_{1}}{2\Delta x_{1}} \right)(y_{1} - z_{1})}{R}} &= e^{2ik_{o}\left(\tilde{n}(y_{1} - z_{1}) + \frac{y_{1}^{2} - z_{1}^{2}}{2\Delta x_{1}} \right)\frac{L_{SA}}{RN}} \\
&= e^{2ik_{o}\tilde{n}(y_{1} - z_{1})\frac{L_{SA}}{RN}}e^{ik_{o}\frac{y_{1}^{2} - z_{1}^{2}}{\Delta x_{1}}\frac{L_{SA}}{RN}} \end{split}.$$

Substituting this expression in for the first term in the summation we see,

$$\begin{split} \sum_{\tilde{n} = -\frac{\tilde{N}}{2}}^{\tilde{n} = \frac{\tilde{N}}{2}}e^{2ik_{o}\frac{x_{1}^{n}(y_{1} - z_{1})}{R}}e^{ik_{o}\frac{z_{1}^{2} - y_{1}^{2}}{R}} &= \sum_{\tilde{n} = -\frac{\tilde{N}}{2}}^{\tilde{n} = \frac{\tilde{N}}{2}} e^{2ik_{o}\frac{\tilde{n}L_{SA}}{RN}(y_{1} - z_{1})} e^{ik_{o}\frac{\left(y_{1}^{2} - z_{1}^{2}\right)}{R}} e^{ik_{o}\frac{\left(z_{1}^{2} - y_{1}^{2} \right)}{R}} \\
&= \sum_{\tilde{n} = -\frac{\tilde{N}}{2}}^{\tilde{n} = \frac{\tilde{N}}{2}} e^{2ik_{o}\frac{\tilde{n}L_{SA}}{RN}(y_{1} - z_{1})} \end{split}$$

Therefore, we see the new expression for $W_{\sum}(\underline{y}, \underline{z})$ is

$$W_{\sum}(\underline{y}, \underline{z}) = e^{-2ik_{o}\frac{Ll}{R}}e^{-2ik_{o}\frac{l^{2}\cos^{2}(\theta)}{R}} \chi_{2L_{SA}}(y_{1} - z_{1}) \sum_{\tilde{n} = -\frac{\tilde{N}}{2}}^{\tilde{n} = \frac{\tilde{N}}{2}}e^{2ik_{o}\frac{\tilde{n}L_{SA}}{RN}(y_{1} - z_{1})}$$

Here, we see that the summation term is a sum of a geometric series defined below where the solution is $S_{n}$ by the first term in the sequence, $a_{1}$, the common ratio, $r$, and $n$ is the number of terms, i.e., $n = q - p$ in the below example:

$$\sum_{j=p}^{q}x^{j}$$
$$a_{1} = x^{p}, \quad r = \frac{a_{n+1}}{a_{n}} = \frac{a_{2}}{a_{1}} = \frac{x^{p+1}}{x^{p}} = x, \quad n = p - q$$
$$S_{n} = \frac{a_{1}(1 - r^{n})}{1 - r} = \frac{x^{p}(1 - x^{n})}{1 - x}$$

In our case, we will get our exponent in the form $e^{i\phi \tilde{n}}:$

$$\phi = 2k_{o}\frac{L_{SA}}{RN}(y_{1} - z_{1})$$
$$\sum_{\tilde{n} = -\frac{\tilde{N}}{2}}^{\tilde{n} = \frac{\tilde{N}}{2}} e^{i\phi \tilde{n}}$$

$$a_{1} = e^{-i\phi \frac{\tilde{N}}{2}}, \quad r = \frac{e^{i\phi \left(-\frac{\tilde{N}}{2} + 1 \right)}}{e^{-i\phi \frac{\tilde{N}}{2}}} = e^{i\phi}, \quad n = \frac{\tilde{N}}{2} - \left(-\frac{\tilde{N}}{2} \right) = \tilde{N}$$

$$\begin{split} S_{n} &= \frac{e^{-i\phi \frac{\tilde{N}}{2}}\left(1 - e^{i\phi \tilde{N}} \right)}{1 - e^{i\phi}} \\
&= \frac{e^{-i\phi \frac{\tilde{N}}{2}} - e^{i\phi \frac{\tilde{N}}{2}} }{e^{i\phi \frac{1}{2}}\left(e^{-i\frac{\phi}{2}} - e^{i\frac{\phi}{2}} \right)}\cdot \left|\frac{e^{i\frac{\phi}{2}}}{e^{i\frac{\phi}{2}}} \right|^{*} \\
&= \frac{-2i\sin\left(\phi\frac{\tilde{N} + 1}{2} \right)}{-2i\sin\left(\frac{\phi}{2} \right)} \\
&= \frac{\sin\left(\phi \frac{\tilde{N} + 1}{2} \right)}{\sin\left(\frac{\phi}{2} \right)} \end{split}$$

$$\sum_{\tilde{n} = -\frac{\tilde{N}}{2}}^{\tilde{n} = \frac{\tilde{N}}{2}} e^{i\phi \tilde{n}} = \frac{\sin\left(\phi \frac{\tilde{N} + 1}{2} \right)}{\sin\left(\frac{\phi}{2} \right)}$$

Where $W_{\sum}(\underline{y}, \underline{z})$ is period with regards to $\sin(\frac{\phi}{2})$ from the sum of the geometric series term and period $2\pi$ and is expressed as,

$$W_{\sum}(\underline{y}, \underline{z}) = e^{-2ik_{o}\frac{Ll}{R}} e^{-ik_{o}\frac{l^{2}\cos^{2}(\theta)}{R}} \chi_{2L_{SA}}(y_{1} - z_{1}) \frac{\sin(\phi \frac{\tilde{N} + 1}{2})}{\sin(\frac{\phi}{2})}.$$

Where the period is,

$$\begin{split} X &= \frac{2\pi}{\phi} \\
&= \frac{2\pi}{k_{o}}\cdot \frac{RN}{L_{SA}} \\
&= \frac{\lambda_{o}RN}{L_{SA}}, \quad \text{where} \quad \lambda_{o} = \frac{2\pi}{k_{o}} \\
&= \frac{\lambda_{o}R}{\Delta x_{1}}, \quad \text{where} \quad \Delta x_{1} = \frac{L_{SA}}{N}. \end{split}$$

Using our previous knowledge of modeling the radiation pattern of the antennas as a $sinc$ function, we know that there will be a main lobe, what we are interested in, and grating lobes, that add undesired affects to the image. We want to try and exclude these grating lobes from our summation term, which we can do if we ensure our periodicity, $X$, is large enough so we can ignore these grating lobes. If $X \gt L_{SA},$ then these grating lobes will not be included in our indicator function $\chi_{2L_{SA}}(y_{1} - z_{1}),$ as we should only be summing up the main lobe due to restricting the period. To ensure this we can utilize the Fraunhofer distance of the synthetic array, $\frac{L_{SA}}{\lambda_{o}}$, which we want to be much larger than $R$ in order to keep the target in the near-field phas distance of at least $\pi$ to allow coherent imaging. And we want $N$ to be very large as well to ensure $X \gt L_{SA}$, so we see:

$$N \gt \frac{L_{SA}}{\lambda_{o}}\cdot \frac{1}{R} \gg 1$$

Additionaly, we can ensure the measure that $X \gt L_{SA}$ by utilizing the speed of the satellite, $v$ (typically $~7.9km/s$), to find some upper bound on the interval between successive emissions, $\tau_{p}$ (typically $~10E-3s$, but in practice it is usually much quicker than this):

$$\Delta x_{1} = v\cdot\tau_{p}, \quad \text{where} \quad \tau_{p} \lt \frac{\lambda_{o}R}{v\cdot L_{SA}}$$

From [2], for the common imaging configurations we have $k_{o}\frac{L_{SA}}{N} = \frac{2\pi v \tau_{p}}{\lambda_{o}} \gt 1$, which indicates that the distance between the successive emitting/receiving locations of the antenna is greater than the carrier wavelength. Therefore, when $\left|y_{1} - z_{1} \right| \lt L_{SA},$ then $|\phi| \underset{\sim}{\lt} 1,$ and $L_{SA} \ll R$ in practice (difference between the synthetic aperture length and the distance from the antenna in orbit to the ground). Then we are interested, mainly, in when $\underline{y}$ & $\underline{z}$ are close to each other (as seen in the previous figure), or $\left|y_{1} - z_{1} \right| \ll L_{SA}.$ Then, we see that $|\phi| \ll 1$ and can approximate $\sin(\frac{\phi}{2}) \approx \frac{\phi}{2}.$ Additionally, the difference between $N$ & $\tilde{N}$ in this interval in small and $N \gg 1$, so we can drop the $+1$ term in the result of our sum of the geometric series and obtain a new approximation of $W_{\sum}(\underline{y}, \underline{z}):$

$$\begin{split} W_{\sum}(\underline{y}, \underline{z}) &= e^{-2ik_{o}\frac{Ll}{R}} e^{-ik_{o}\frac{l^{2}\cos^{2}(\theta)}{R}} \chi_{2L_{SA}}(y_{1} - z_{1}) \frac{\sin(\phi \frac{\tilde{N}}{2})}{\frac{\phi}{2}} \\
&= e^{-2ik_{o}\frac{Ll}{R}} e^{-ik_{o}\frac{l^{2}\cos^{2}(\theta)}{R}} \chi_{2L_{SA}}(y_{1} - z_{1}) \tilde{N}\frac{\sin(\phi \frac{\tilde{N}}{2})}{\tilde{N}\frac{\phi}{2}} \\
&= e^{-2ik_{o}\frac{Ll}{R}} e^{-ik_{o}\frac{l^{2}\cos^{2}(\theta)}{R}} \chi_{2L_{SA}}(y_{1} - z_{1}) \tilde{N}sinc\left(\frac{\tilde{N}\phi}{2} \right). \end{split}$$

Now, we are in a position to examine the azimuthal resolution, $\Delta_{A}$, which was our original objective. This can be discovered by examining the $sinc$ portion of the above expression and identifying the distance of the main lobe.

$$\begin{split} sinc\left(\frac{\tilde{N}\phi}{2} \right) &= sinc\left(\frac{\tilde{N}}{2}2k_{o}\frac{L_{SA}}{RN}(y_{1} - z_{1})\right) \\
&= sinc\left(k_{o}(y_{1} - z_{1})\frac{L_{SA}}{R}\frac{\tilde{N}}{N} \right) \\
&= sinc\left(k_{o}(y_{1} - z_{1})\frac{L_{SA}}{R}\left(1 - \frac{\left|y_{1} - z_{1} \right|}{N\Delta x_{1}} \right) \right), \quad \text{where} \quad \tilde{N} = N - \left[\frac{\left|y_{1} - z_{1} \right|}{\Delta x_{1}} \right] \end{split}$$

We are interested in the first zeros (left & right) of the main lobe of this $sinc$ function, which will occur at $sinc(\pi)$, therefore, we will equate the above expression to $\pi$ and solve for $(y_{1} - z_{1})$, as we can see this is quadratic with respect to $(y_{1} - z_{1})$. One of our $(y_{1} - z_{1})$ is an absolute value, we will take examine when this value is $-$, as the result when $+$ is very similar.

$$\begin{split} \frac{k_{o}L_{SA}}{R}\left(\frac{\left(y_{1} - z_{1} \right)^{2}}{N\Delta x_{1}} + (y_{1} - z_{1}) \right) &= \pi \\
(y_{1} - z_{1})^{2}\frac{k_{o}L_{SA}}{RN\Delta x_{1}} + (y_{1} - z_{1})\frac{k_{o}L_{SA}}{R} - \pi &= 0\end{split}$$

Solving, using the quadratic formula:

$$x = \frac{-b \pm \sqrt{b^{2} - 4ac}}{2a}$$
$$x = (y_{1} - z_{1}), \quad a = \frac{k_{o}L_{SA}}{RN\Delta x_{1}}, \quad b = \frac{k_{o}L_{SA}}{R}, \quad c = -\pi$$

$$\begin{split} (y_{1} - z_{1}) &= \frac{\frac{-k_{o}L_{SA}}{R} \pm \sqrt{\frac{k_{o}^{2}L_{SA}^{2}}{R^{2}} + 4\frac{k_{o}\pi L_{SA}}{RN\Delta x_{1}}}}{2\frac{k_{o}L_{SA}}{RN\Delta x_{1}}} \\
&= \frac{-k_{o}\frac{L_{SA}}{R} \pm \sqrt{\frac{k_{o}^{2}L_{SA}^{2}}{R^{2}}\left(1 + 4\frac{R\pi}{N\Delta x_{1}L_{SA}k_{o}} \right) } }{2\frac{k_{o}L_{SA}}{RN\Delta x_{1}}} \\ 
&= \frac{-k_{o}\frac{L_{SA}}{R} \pm k_{o}\frac{L_{SA}}{R}\sqrt{1 + \frac{4\pi R}{L_{SA}^{2}k_{o}}} }{2\frac{k_{o}L_{SA}}{RL_{SA}}} \\
&= \frac{R}{2k_{o}}\left(-k_{o}\frac{L_{SA}}{R} \pm k_{o}\frac{L_{SA}}{R}\sqrt{1 + \frac{4\pi R}{L_{SA}^{2}k_{o}}} \right). \end{split}$$

Since the non-trivial term under the square root contains the inverse Fraunhofer distance multiplied against R, and we know from earlier that we are only interested in when the Fraunhofer distance is much larger than the distance to the ground, $\frac{L_{SA}}{\lambda_{o}} \gg R$, to perform coherent imaging, this term will be very small and we can approximate it via the Taylor series:

$$\begin{split} \sqrt{1 + \frac{4\pi R}{L_{SA}^{2}k_{o}}} &\approx 1 + \frac{1}{2}\frac{4\pi R}{L_{SA}^{2}k_{o}} - \frac{1}{8}\frac{16\pi^{2} R^{2}}{L_{SA}^{4}k_{o}^2} \\
&\approx 1 + \frac{2\pi R}{L_{SA}^{2}k_{o}}, \quad \text{where} \quad \frac{1}{8}\frac{16\pi^{2} R^{2}}{L_{SA}^{4}k_{o}^2} \to 0 \end{split}$$

$$(y_{1} - z_{1}) = \frac{R}{2k_{o}}\left(-k_{o}\frac{L_{SA}}{R} \pm k_{o}\frac{L_{SA}}{R}\left(1 + \frac{2\pi R}{L_{SA}^{2}k_{o}} \right) \right)$$

Analyzing the first root, $+$:

$$\begin{split} (y_{1} - z_{1}) &= -\frac{L_{SA}}{2} + \frac{L_{SA}}{2}\left(1 + \frac{2\pi R}{L_{SA}^{2}k_{o}} \right) \\
&= -\frac{L_{SA}}{2} + \frac{L_{SA}}{2} + \frac{\pi R}{L_{SA}k_{o}} \\
&= \frac{\pi R}{L_{SA}k_{o}} \\
&= \frac{\pi Rc}{L_{SA}\omega_{o}}, \quad \text{where} \quad k_{o} = \frac{\omega_{o}}{c} \\
&= \Delta_{A} \end{split}.$$

We have arrived at our azimuthal resolution. Now, if the length of the synthetic aperture $L_{SA}$ is chosen equal to the size of the antenna footprint on the ground, we see, from [2, equation (2.29)] that $L_{SA} = 2\frac{\lambda_{o}}{D}R$. Substituting this value into our above expression, we see:

$$\begin{split} \Delta_{A} &= \frac{\pi Rc}{\omega_{o}L_{SA}} \\
&= \frac{\pi Rc}{\omega_{o}\left(\frac{2R\lambda_{o}}{D} \right)} \\
&= \frac{\pi Rc}{\omega_{o}\left(\frac{2R}{\omega_{o}D} \right)} \\
&= \frac{\pi cD}{2}. \end{split}$$

And we can clearly see that the azimuthal resolution depends neither on the carrier frequency $\omega_{o}$ nor distance to the target $R$. We can also see the following relationship,

$$\Delta_{A} \propto D$$

therefore, we can see that as the antenna size, $D$, gets smaller, as will the azimuthal resolution. Since resolution is defined as the distance able to differentiate between multiple point scatterers, we would want the main lobe of our $sinc$ function to be as narrow as possible, which is expressed as $\Delta_{A}$, in order to differentiate between multiple peaks that are close to each other. I.e., we want to approach a $\delta$ function as closely as possible, which we do as $D$ decreases and, consequently, $\Delta_{A}$ gets smaller.

Consider the point spread function (PSF) given by [2, equation (2.24)]:

$$W_{x}(\underline{y}, \underline{z}) = \int_{\chi}\overline{P\left(t - \frac{2R_{\underline{y}}}{c} \right)} P\left(t - \frac{2R_{\underline{z}}}{c} \right)dt.$$

Show that, on the interval $\omega \in \left[\omega_{o} - \frac{B}{2}, \omega_{o} + \frac{B}{2} \right]$ the spectrum of the PSF $W_{x}$ coincides with the spectrum of the delta function $\delta\left(\frac{2R_{\underline{y}}}{c} - \frac{2R_{\underline{z}}}{c} \right),$ up to a constant factor:

$$W_{x}(\underline{y}, \underline{z}) = \frac{\pi}{\alpha}\int_{\omega_{o} - \frac{B}{2}}^{\omega_{o} + \frac{B}{2}} e^{-i\omega\left(\frac{2R_{\underline{y}}}{c} - \frac{2R_{\underline{z}}}{c} \right)}d\omega,$$

whereas

$$\delta\left(\frac{2R_{\underline{y}}}{c} - \frac{2R_{\underline{z}}}{c} \right) = \frac{1}{2\pi}\int_{-\infty}^{\infty} e^{-i\omega\left(\frac{2R_{\underline{y}}}{c} - \frac{2R_{\underline{z}}}{c} \right)}d\omega.$$

This is another way to understand in what sense the PSF $W_{x}$ approximates $\delta\left(\frac{2R_{\underline{y}}}{c} - \frac{2R_{\underline{z}}}{c} \right).$ Note that, in the spatially one-dimensional case: $\underline{z} = z = R_{\underline{z}}$ and $\underline{y} = y = R_{\underline{y}},$ we have $\delta(y - z) = \frac{2}{c}\delta\left(\frac{2R_{\underline{y}}}{c} - \frac{2R_{\underline{z}}}{c} \right).$ See the analysis in [2, Appendix 2.A]. In particular, you can use [2, equation (2.108)]. Also see the discussion in [2, Section 2.4.5].

Let the received field by the array be

$$u^{(1)}(t, \underline{x}) = \int \nu(\underline{z})P\left(t - 2\frac{R_{\underline{z}}}{c} \right)d\underline{z},$$

where $\nu(\underline{z})$ is the ground reflectivity and propagation attenutation. Allow the Image, $I_{x}(\underline{y})$, be defined as:

$$\begin{split} I_{x}(\underline{y}) &= \int_{-\infty}^{\infty}K(t, \underline{y})u^{(1)}(t, \underline{x})dt \\
&= \int \nu(\underline{z})\int_{-\infty}^{\infty}K(t, \underline{y})P\left(t - 2\frac{R_{\underline{z}}}{c} \right)dtd\underline{z}. \end{split}$$

Where $K(t, \underline{y})$ is a helper function, that can be used to help in our reconstruction of $\nu(\underline{z})$ as the image $I_{x}(\underline{y}).$ And we will define $W_{x}(\underline{y}, \underline{z}) = \int_{-\infty}^{\infty}K(t, \underline{y})P\left(t - 2\frac{R_{\underline{z}}}{c} \right)dt.$

Now, we notice that the inner integral is closely related to our PSF. We also want our image to be as close to $\nu(\underline{z})$ as possible, in order to reconstruct the scene in our image as closely as possible. This can be accomplished if $\int_{-\infty}^{\infty}K(t, \underline{y})P\left(t - 2\frac{R_{\underline{z}}}{c} \right)dt$ is the dirac-delta function, $\delta(\underline{y} - \underline{z})$, or as close to it as possible. We need to figure out a way to compare the closeness of the PSF function to the dirac-delta function, however, due to Kolmogorov's normability criterion, where a space is normable if and only if only finitely many of the spaces are non-trivial, $\delta(\underline{y} - \underline{z})$ is not inheritantly equipped with a norm in order to allow us to do this. In this case, we can use the spectral theorem to transform our PSF and dirac-delta function in order to transform the PSF and the dirac-delta function onto a space where they can be suitably compared.

Using the Fourier transform, let $\tilde{K}$ & $\tilde{P}$ be the Fourier transforms of K & P, respectively, in time,

$$\begin{split} \tilde{K}(\omega, \underline{y}) &= \int_{-\infty}^{\infty}K(t, \underline{y})e^{i\omega t}dt \\
\tilde{P}(\omega) &= \int_{-\infty}^{\infty}P(t)e^{i\omega t}dt \\
&= \int_{-\infty}^{\infty}P\left(t - 2\frac{R_{\underline{z}}}{c}\right)e^{i\omega\left(t - 2\frac{R_{\underline{z}}}{c} \right)}dt \\
&= \int_{-\infty}^{\infty}e^{i\omega 2\frac{2R_{\underline{z}}}{c}}P\left(t - 2\frac{R_{\underline{z}}}{c} \right) e^{i\omega t}dt. \end{split}$$

Since the Fourier transform of a product is the convolution of Fourier transforms, we can rewrite $W_{x}(\underline{y}, \underline{z})$ as:

$$W_{x}(\underline{y}, \underline{z}) = \frac{1}{2\pi}\int_{-\infty}^{\infty}e^{i\omega 2\frac{R_{\underline{z}}}{c}} \tilde{P}(\omega) \tilde{K}(-\omega, \underline{y})d\omega.$$

Using the spatially one-dimensional case, as outlined in the question, where $\underline{z} = R_{\underline{z}} \equiv z$ & $\underline{y} = R_{\underline{y}} \equiv y,$ then the Fourier transform of $\delta(y - z)$ becomes:

$$\delta(y - z) = \frac{2}{c}\delta(2\frac{R_{y}}{c} - 2\frac{R_{z}}{c}) = \frac{2}{c}\frac{1}{2\pi}\int_{-\infty}^{\infty}e^{-i\omega 2\frac{R_{y} - R_{z}}{c}}d\omega.$$

Equating the values under the integral of $W_{x}(\underline{y}, \underline{z})$ and under the integral of $\delta(y - z)$, alongside their constants of $\frac{1}{2\pi}$, we get:

$$\begin{split} \frac{1}{2\pi} e^{i\omega 2\frac{R_{z}}{c}}\tilde{P}(\omega)\tilde{K}(-\omega, \underline{y}) &= \frac{1}{2\pi} e^{-i\omega 2\frac{R_{y} - R_{z}}{c}} \\
e^{i\omega 2\frac{R_{z}}{c}}\tilde{P}(\omega)\tilde{K}(-\omega, \underline{y}) &= e^{-i\omega 2\frac{R_{y}}{c}}e^{i\omega 2\frac{R_{z}}{c}} \\
\tilde{P}(\omega)\tilde{K}(-\omega, \underline{y}) &= e^{-i\omega 2\frac{R_{y}}{c}}. \end{split}$$

Here, we can see that the expressions only differ by a constant of $\frac{2}{c}.$ If the above expression could be satisfied for all $\omega$, then $W_{x}(y, z)$ would be proprtional to $\delta(y - z),$ which would give us back our $\nu(z)$ expression as our image, $I_{x}(y)$. Using the definition of $\tilde{P}(\omega)$ from [2, equation (2.108)],

$$\begin{split} \tilde{P}(\omega) &= \chi_{\tau}(t) e^{-i\alpha t^{2}}e^{-i\omega_{0}t} e^{i\omega t}dt \\
&\approx \sqrt{\frac{\pi}{\alpha}}e^{-i\frac{\pi}{4}} e^{i\frac{(\omega - \omega_{0})^{2}}{4\alpha}}\chi_{B}(\omega - \omega_{0}). \end{split}$$

Where, the chirped signal is designed to have its spectrum confined to a band of width B around the central carrier frequency $\omega_{0},$ defined by:

$$\chi_{B}(\omega - \omega_{0}) = \begin{cases} 1, & \left|\omega - \omega_{0} \right| \leq \frac{B}{2} \\ 0, & \text{otherwise}. \end{cases}$$

Which originates from the condition that the stationary point of the phase must belong to the interval $\left[\frac{-\tau}{2}, \frac{\tau}{2} \right]$ defined by the indicator function $\chi_{\tau}(t)$. Therefore, $\tilde{K}(\omega - \omega_{0})$ to satisfy the equality to the Fourier transform of the dirac-delta function above is,

$$\tilde{K}(\omega, y) = \frac{1}{\tilde{P}(\omega)}e^{i\omega 2\frac{R_{y}}{c}} = \text{const} \cdot \tilde{\overline{P}}(\omega)e^{i\omega 2\frac{R_{y}}{c}},$$

where $\tilde{\overline{P}}(\omega)$ is the Fourier transform of the complex conjugate of $P(t)$, and we can see that $K(t, y)$ is directly proportional to $\bar{P}$:

$$K(t, y) \propto \overline{P\left(t - 2\frac{R_{y}}{c} \right)}.$$

Therefore,

$$\begin{split} W_{x}(y, z) &= \int_{-\infty}^{\infty}K(t, y)P\left(t - 2\frac{R_{z}}{c} \right)dt \\
&= \text{const}\cdot\int_{-\infty}^{\infty}\overline{P\left(t - 2\frac{R_{y}}{c} \right)}P\left(t - 2\frac{R_{z}}{c} \right)\chi_{B}(\omega - \omega_{0})dt \\
&= \text{const}\cdot\int_{\chi_{B}}\overline{P\left(t - 2\frac{R_{y}}{c} \right)}P\left(t - 2\frac{R_{z}}{c} \right)dt \end{split}$$

is equivalent to our original expression by some constant differentiating our original PSF and the PSF derived to be as close as possible to the dirac-delta function.

&= -\Delta_{R} - \frac{\tau}{2}. \end{split}$$

The $+$ root, $-\Delta_{R} - \frac{\tau}{2},$ is not of interest in this case of analyzing the main lobe of the $sinc$ function, because $|T^{c}| \sim \tau$ as opposed to $|T^{c}| \ll \tau.$ So we are interested in $\Delta_{R}$ root.

We can now see that the main lobe of the $sinc$ function in $W_{R}$, when considered a function of $T^{c}$, is located on the interval $\left[-\Delta_{R}, \Delta_{R} \right].$

From the Cauchy-Schwarz inequality, we know the following statements:

$$\left|\left<x, y\right>\right|^{2} \leq \left<x, x\right>\left<y, y\right>$$
$$\left|\left<x, y \right> \right|^{2} = \left<x, x\right>\left<y, y\right> \text{  holds iff  } y=ax \text{  for some  } a \in C \text{. That is, equality holds iff  } x \text{  and  }  y \text{  are linearly dependent.}$$

Where $\left<\cdot, \cdot \right>$ indicates the inner product & $C$ is the set of complex numbers. $$\int_{-\infty}^{\infty}\left|K(t, \underline{y}) \right|^{2}dt \leq E_{K}.$$
