Since we are taking into account the movement of the platform during the transmission of the radar signals emitted, we will utilize the Lorentz transform in order to transform a moving object to a coordinate system that is stationary, while keeping the wave equation and Maxwell's equations invariant. Thus, we will have $\left(t, z_{1}, z_{2}, z_{3} \right) \mapsto \left(\sigma, \zeta_{1}, z_{2}, z_{3} \right)$. Where,

$$\sigma = \frac{1}{\beta}\left(t - \frac{vz_{1}}{c^{2}} \right), \quad \zeta_{1} = \frac{1}{\beta}\left(-vt + z_{1} \right),$$
and
$$\beta = \sqrt{1 - \frac{v^{2}}{c^{2}}}.$$

Thus, leading to the following wave equation:

$$\begin{split} \frac{1}{c^{2}}\frac{\partial^{2}u}{\partial\sigma^{2}} - \frac{\partial^{2}u}{\partial\zeta_{1}^{2}} - \frac{\partial^{2}u}{\partial z_{2}^{2}} - \frac{\partial^{2}u}{\partial z_{3}^{2}} &= P\left(\frac{\sigma}{\beta} + \frac{\frac{v}{c}}{\beta}\frac{\zeta_{1}}{c} \right)\delta\left(\beta\zeta_{1} \right)\delta\left(z_{2} + L \right)\delta\left(z_{3} - H \right) \\
&= f\left(\sigma, \zeta_{1}, z_{2}, z_{3} \right). \end{split}$$

Thus, using Kirchhoff's integral theorem to define the impinging field, we get

$$u^{(0)}\left(\sigma, \zeta_{1}, z_{2}, z_{3} \right) = \frac{1}{4\pi}\int\int\int\frac{f\left(\sigma - \frac{\rho'}{c}, \zeta_{1}', z_{2}', z_{3}' \right)}{\rho'}d\zeta_{1}'dz_{2}'dz_{3}',$$
where
$$\rho' = \left(\left(\zeta_{1} - \zeta_{1}' \right)^{2} + \left(z_{2} - z_{2}' \right)^{2} + \left(z_{3} - z_{3}' \right)^{2} \right)^{\frac{1}{2}}.$$

Thus, defining $\zeta_{1}' = 0$, as we have transformed into a coordinate system where our antenna is no longer moving, $z_{2}' = -L$, as this is the projection distance of the antenna, $z_{3}' = H$, as this is the height of the antenna from the ground. Thus, we are saying that the antenna is at rest at $\boldsymbol{x} = (0, -L, H)$. Integrating $u^{(0)}$ we get,

$$u^{(0)} = \frac{1}{4\pi\beta}\frac{P\left(\frac{\sigma - \frac{\rho}{c}}{\beta} \right)}{\rho},$$
where
$$\rho = \left(\zeta_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2} \right)^{\frac{1}{2}}.$$

Notice, if we drop $\beta$, then we get the straightforward retarded potential in the new coordinate system. Solving for $\rho$,

$$\rho = \left(\zeta_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2} \right)^{\frac{1}{2}}$$

$$\begin{split} \zeta_{1}^{2} &= \left(\frac{1}{\beta}\left(-vt + z_{1} \right) \right)^{2} \\
&= \frac{1}{\beta^{2}}\left(z_{1} - vt \right)^{2} \\
&= \frac{1}{\beta^{2}}\left(z_{1}^{2} - 2z_{1}vt + \left(vt \right)^{2} \right) \\
&= \frac{z_{1}^{2}}{\beta^{2}} + \frac{1}{\beta^{2}}\left(-2z_{1}vt + \left(vt \right)^{2} \right) \end{split},$$
where
$$\begin{split} \frac{z_{1}^{2}}{\beta^{2}} &= z_{1}^{2} - z_{1}^{2} + \frac{z_{1}^{2}}{\beta^{2}} \\
&= z_{1}^{2} + z_{1}^{2}\left(\frac{1}{\beta^{2}} - 1 \right) \\
&= z_{1}^{2} + \frac{1 - \beta^{2}}{\beta^{2}}z_{1}^{2} \\
&= z_{1}^{2}. \end{split}$$

Where we dropped any $\frac{c^{2}}{v^{2}} \gg 1$ terms. Thus,

$$\begin{split} \zeta_{1}^{2} &= z_{1}^{2} + \frac{1}{\beta^{2}}\left(-2z_{1}vt + \left(vt \right)^{2} \right) \\
&= z_{1}^{2} - 2z_{1}vt + \left(vt \right)^{2} + \frac{1 - \beta^{2}}{\beta^{2}}\left(z_{1} - vt \right)^{2} \\
&= z_{1}^{2} - 2z_{1}vt + \left(vt \right)^{2} \end{split}$$

$$\rho \approx \left(z_{1}^{2} - 2z_{1}vt + \left(vt \right)^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2} \right)^{\frac{1}{2}}$$

$$r = R_{z}\bigg\rvert_{t=0,z_{3}=0} = \sqrt{z_{1}^{2} + \left(z_{2} + L \right)^{2} + H^{2}}.$$

Where $r$ defines the moment of time $t = 0$ when the center of the pulse is emitted corresponds to the location $\boldsymbol{x}(0) = (0, -L, H)$ on the orbit. Therefore, when $z_{3} = 0$,

$$\rho \approx \left(r^{2} - 2z_{1}vt + \left(vt \right)^{2} \right)^{\frac{1}{2}} .$$

Furthermore, the timescale we are interested in is the travel time of the impinging signal to propagate to the target or $t \sim \frac{r}{c}$. Therefore, the 3rd term on the right-hand side is only taken into account if we have a very narrow sub-angle or $\left|z_{1}\right|\left|vt\right| \sim \left(vt \right)^{2}$ as $\left(vt \right)^{2} = \left(v\frac{r}{c} \right)^{2} = \left(\frac{v}{c}r \right)^{2} = \frac{v^{2}}{c^{2}}r^{2}$, so we would assume to drop this term as well due to $\frac{v^{2}}{c^{2}}$. Note, if $v = 0$, we get back the solution using the Start-Stop approximation. Now, we need to perform another Lorentz transform on the received, scattered, field at the moving antenna receiver. Therefore, we must perform the mapping $(t, x_{1}', x_{2}', x_{3}') \mapsto (\vartheta, \xi_{1}, x_{2}', x_{3}')$, where

$$\vartheta = \frac{1}{\beta}\left(t - \frac{vx_{1}'}{c^{2}} \right), \quad \xi_{1} = \frac{1}{\beta}\left(-vt + x_{1}' \right).$$

Note, $t - \frac{r}{c}$ is the moment the pulse is reflected at the target, defining $\breve{t} = \frac{r}{c}$. Therefore, the moment the emitting pulse is re-emitted at the target $\boldsymbol{z}$, event of reflection, is

$$\breve{\vartheta} = \frac{1}{\beta}\left(\breve{t} - \frac{vz_{1}}{c} \right), \quad \breve{\xi}_{1} = \frac{1}{\beta}\left(-v\breve{t} + z_{1} \right).$$

We also know, from derivations in earlier homeworks, that the maximum of the $sinc$ argument in the impinging field, $u^{(0)}\left(t, \boldsymbol{z}\right),$ definition is $\frac{\lambda_{0}}{D}.$ Where $\lambda_{0}$ is the wavelength of the emitting radiation pattern and $D$ is the length the of synthetic aperture. Thus the upper bound on the emitting angle is $\frac{\lambda_{0}}{D}$, where this produces the maximum level of radiation emitted by a linear antenna, when parallel to the orbit. Henceforth, the condition where the moment of reflection of the point $\boldsymbol{z}$ is within the antenna beam is expressed as:

$$\frac{\left|\breve{\xi_{1}} \right|}{R} \leq \frac{\lambda_{0}}{D}.$$

For $t \gt \breve{t} = \frac{r}{c}$, the scattered field is obtained by:

$$u^{(1)}\left(t, \boldsymbol{x}'\right) \approx \int \nu(\boldsymbol{z}) P\left(\left(t - \breve{t} - \frac{r'}{c} \right) \left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{\beta R} \right)d\boldsymbol{z}.$$

Where $t - \breve{t} - \frac{r'}{c}$ comes from the event of reflection at the target, $\breve{t},$ and the modified distance between the moving antenna, $\boldsymbol{x}'$, and the target, $\boldsymbol{z}.$ Which is a doppler shift of the retarded time. Where,

$$\begin{split} r' &= \sqrt{\left(x_{1}' - z_{1} \right)^{2} + \left(z_{2} + L \right)^{2} + H^{2}} \\
&= \left|\boldsymbol{x}' - \boldsymbol{z} \right|. \end{split}$$

The indicator function, $\chi_{\Theta}$ comes from $\frac{\left|\breve{\xi}_{1} \right|}{R} \leq \frac{\lambda_{0}}{D},$ where this is the permissible range of the angle of the emitting radiation.

Performing the inverse transform of $\vartheta$ & $\xi_{1}$, we get the following expressions:

$$t = \frac{1}{\beta}\left(\vartheta + \frac{v\xi_{1}}{c^{2}} \right), \quad x_{1}' = \frac{1}{\beta}\left(v\vartheta + \xi_{1} \right).$$

Thus, substituting all values into $t - \breve{t} - \frac{r'}{c}$:

$$t - \breve{t} - \frac{r'}{c} = \frac{1}{\beta}\left(\vartheta + \frac{v\xi_{1}}{c^{2}} \right) - \frac{1}{\beta}\left(\breve{\vartheta} + \frac{c\breve{\xi_{1}}}{c^{2}} \right) - \frac{1}{c}\sqrt{\left(\frac{1}{\beta}\left(v\vartheta + \xi_{1} \right) - z_{1} \right)^{2} + \left(z_{2} + L \right)^{2} + H^{2}}.$$

Then, taking $\beta \approx 1$ and finding the inverse of $z_{1}$ with regards to $\breve{\vartheta}$ & $\breve{\xi_{1}}$, we get

$$t - \breve{t} - \frac{r'}{c} \approx \left(\vartheta - \breve{\vartheta} \right) + \frac{v}{c^{2}}\left(\xi_{1} - \breve{\xi_{1}} \right) - \frac{1}{c}\sqrt{\left(\xi_{1} - \left(\breve{\xi_{1}} + v\breve{\vartheta} \right) + v\vartheta \right)^{2} + \left(z_{2} + L \right)^{2} + H^{2}}.$$

Solving for $\left(\xi_{1} - \left(\breve{\xi_{1}} + v\breve{\vartheta} \right) + v\vartheta \right)^{2}$:

$$\begin{split} \left(\xi_{1} - \left(\breve{\xi_{1}} + v\breve{\vartheta} \right) + v\vartheta \right)^{2} &= \left(\xi_{1} - \breve{\xi_{1}} - v\breve{\vartheta} + v\vartheta \right)^{2} \\
&= \xi_{1}^{2} - \xi_{1}\breve{\xi_{1}} - \xi_{1}v\breve{\vartheta} + \xi_{1}v\vartheta - \xi_{1}\breve{\xi_{1}} + \breve{\xi_{1}}^{2} + \breve{\xi_{1}}v\breve{\vartheta} - \breve{\xi_{1}}v\vartheta - \xi_{1}v\breve{\vartheta} + \breve{\xi_{1}}v\breve{\vartheta} + \left(v\breve{\vartheta} \right)^{2} - v^{2}\breve{\vartheta}\vartheta + \xi_{1}v\vartheta - \breve{\xi_{1}}v\vartheta - v^{2}\breve{\vartheta}\vartheta + \left(v\vartheta \right)^{2} \\

&= \left(\xi_{1} - \breve{\xi_{1}} \right)^{2} + 2v\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right) + v^{2}\left(\vartheta - \breve{\vartheta} \right)^{2} \end{split}.$$

Therefore,

$$t - \breve{t} - \frac{r'}{c} = \left(\vartheta - \breve{\vartheta} \right) + \frac{v}{c^{2}}\left(\xi_{1} - \breve{\xi_{1}} \right) - \frac{1}{c}\sqrt{\left(\xi_{1} - \breve{\xi_{1}} \right)^{2} + 2v\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right) + v^{2}\left(\vartheta - \breve{\vartheta} \right)^{2} + \left(z_{2} + L \right)^{2} + H^{2}}.$$

Introducing, $\varrho = \sqrt{\left(\xi_{1} - \breve{\xi_{1}} \right)^{2} + \left(z_{2} + L \right)^{2} + H^{2} }$

$$t - \breve{t} - \frac{r'}{c} = \left(\vartheta - \breve{\vartheta} \right) + \frac{v}{c^{2}}\left(\xi_{1} - \breve{\xi_{1}} \right) - \frac{1}{c}\sqrt{\varrho^{2} + 2v\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right) + v^{2}\left(\vartheta - \breve{\vartheta} \right)^{2} }.$$

Analyzing the third term containing the square root,

$$\frac{1}{c}\sqrt{\varrho^{2} + 2v\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right) + v^{2}\left(\vartheta - \breve{\vartheta} \right)^{2}} = \sqrt{\frac{\varrho^{2}}{c^{2}} + \frac{2v}{c^{2}}\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right) + \frac{v^{2}}{c^{2}}\left(\vartheta - \breve{\vartheta} \right)^{2} }.$$

The $\frac{v^{2}}{c^{2}}$ term is approximately zero. Therefore,

$$\frac{1}{c}\sqrt{\varrho^{2} + 2v\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right) + v^{2}\left(\vartheta - \breve{\vartheta} \right)^{2}} \approx \frac{1}{c}\sqrt{\varrho^{2} + 2v\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right)}.$$

Applying the Taylor series to the expression under the square root where $\alpha = \frac{2v}{\varrho^{2}}\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right)$ and is sufficiently small, so $\sqrt{1 + \alpha} \approx 1 + \frac{1}{2}\alpha - \frac{1}{4}\alpha^{2} + \cdots$ and taking the first two terms:

$$\begin{split} \frac{1}{c}\sqrt{\varrho^{2} + 2v\left(\vartheta - \breve{\vartheta} \right)\left(\xi_{1} - \breve{\xi_{1}} \right) + v^{2}\left(\vartheta - \breve{\vartheta} \right)^{2}} &\approx \frac{\varrho}{c} + \frac{\varrho}{c}\frac{1}{2}\left(\frac{2v}{\varrho^{2}}\left(\vartheta - \breve{\vartheta} \right) \left(\xi_{1} - \breve{\xi_{1}} \right) \right) \\
&= \frac{\varrho}{c} + v\left(\vartheta - \breve{\vartheta} \right)\frac{1}{c}\frac{\xi_{1} - \breve{\xi_{1}}}{\varrho}. \end{split}$$

Thus,

$$\begin{split} t - \breve{t} - \frac{r'}{c} &\approx \left(\vartheta - \breve{\vartheta} \right) + \frac{v}{c^{2}}\left(\xi_{1} - \breve{\xi_{1}} \right) - \frac{\varrho}{c} - v\left(\vartheta - \breve{\vartheta} \right)\frac{1}{c}\frac{\xi_{1} - \breve{\xi_{1}}}{\varrho} \\
&= \left(\vartheta - \breve{\vartheta} - \frac{\varrho}{c} \right)\left(1 + \frac{v}{c}\frac{\breve{\xi_{1}} - \xi_{1}}{\varrho} \right). \end{split}$$

The quantity $\frac{\breve{\xi_{1}} - \xi_{1}}{\varrho}$ can be interpreted as cosine of the angle between the velocity and the direction of the observation point, where the scattered field is received on the antenna, to the pulse re-emission location $\breve{\xi_{1}}$ at the re-emission time $\breve{\vartheta}$, thus,

$$\frac{\breve{\xi_{1}} - \xi_{1}}{\varrho} = \cos(\gamma_{\boldsymbol{z}}').$$

Therefore,

$$t - \breve{t} - \frac{r'}{c} = \left(\vartheta - \breve{\vartheta} - \frac{\varrho}{c} \right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}') \right).$$

Thus, substituting the above expression into the defintion of $u^{(1)}$ we get:

$$u^{(1)}\left(t, \boldsymbol{x}' \right) \approx \int\nu(\boldsymbol{z})P\left(\left(\vartheta - \breve{\vartheta} - \frac{\varrho}{c} \right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}') \right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right)d\boldsymbol{z}.$$

The overall Doppler effect for the scattered field received at the antenna is a combination of two contributions. One, where the field is emitted by a moving antenna onto a motionless target, which is given by $1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}})$. The second, where the impinging field is scattered by a motionless target onto a moving antenna, which is given by $1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}')$. Expanding our expression for  $u^{(1)}$ and neglecting the $\frac{v^{2}}{c^{2}}$ terms, we get:

$$u^{(1)}(t, \boldsymbol{x}') \approx \int \nu(\boldsymbol{z})P\left(\left(\vartheta - \breve{\vartheta} - \frac{\varrho}{c} \right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}') + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right)d\boldsymbol{z}.$$

This expression of the scattered field is still in the Lorentz-transformed coordinates, so we will want to convert the coordinate system back into our original cartesian coordinate system,

$$u^{(1)}(t, \boldsymbol{x}') \approx \int \nu(\boldsymbol{z})P\left(\left(t - \frac{vx_{1}'}{c^{2}} - \frac{r}{c} + \frac{vz_{1}}{c^{2}} - \frac{\varrho}{c} \right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}') + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right)d\boldsymbol{z}.$$

When the scattered signal is received at the antenna, $\xi_{1} = 0$, therefore,

$$\begin{split} \varrho &= \sqrt{\breve{\xi_{1}}^{2} + \left(z_{2} + L \right)^{2} + H^{2}} \\
&= \sqrt{\left(z_{1} - \frac{v}{c}r \right)^{2} + \left(z_{2} + L \right)^{2} + H^{2}} \\
&= \sqrt{z_{1}^{2} - 2\frac{v}{c}z_{1}r + \frac{v^{2}}{c^{2}}r^{2} + \left(z_{2} + L \right)^{2} + H^{2}} \\
&= \sqrt{z_{1}^{2} - 2\frac{v}{c}z_{1}r + \left(z_{2} + L \right)^{2} + H^{2}} \\
&= \sqrt{r^{2} - 2z_{1}\frac{v}{c}r} \\
&= r\sqrt{1 - 2z_{1}\frac{v}{cr}} \\
&\approx r\left(1 - \frac{1}{2}\left(2z_{1}\frac{v}{cr} \right) + \frac{1}{4}\left(2z_{1}\frac{v}{cr} \right)^{2} \right) \\
&\approx r - z_{1}\frac{v}{c}, \end{split}$$

where we neglected the $\frac{v^{2}}{c^{2}}$ terms. Substituting this expression into $\cos(\gamma_{\boldsymbol{z}}')$ at $\xi_{1} = 0$ and the inverse of $\breve{\xi_{1}}$, we get

$$\begin{split} \cos(\gamma_{\boldsymbol{z}}') = \frac{\breve{\xi_{1}} - \xi_{1}}{\varrho}\bigg\rvert_{\xi_{1} = 0} &\approx \frac{z_{1} - \frac{v}{c}r}{r - z_{1}\frac{v}{c}} \\
&\approx \frac{z_{1} - \frac{v}{c}r}{r\left(1 - \frac{z_{1}}{r}\frac{v}{c} \right)} \\
&\approx \frac{z_{1} - \frac{v}{c}r}{r}\left(1 + \frac{z_{1}}{r}\frac{v}{c} \right) \\
&= \frac{z_{1} - \frac{v}{c}r}{r}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}})\right) \\
&= \left(\cos(\gamma_{\boldsymbol{z}}) - \frac{v}{c} \right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right). \end{split}$$

Henceforth,

$$\begin{split} 1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}') + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) &= 1 + \frac{v}{c}\left(\cos(\gamma_{\boldsymbol{z}}) - \frac{v}{c} \right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \\
&= 1 + \frac{v}{c}\left(\cos(\gamma_{\boldsymbol{z}}) + \frac{v}{c}\cos^{2}(\gamma_{\boldsymbol{z}}) - \frac{v}{c} - \frac{v^{2}}{c^{2}}\cos(\gamma_{\boldsymbol{z}}) \right) + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \\
&\approx 1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \\
&= 1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}). \end{split}$$

Substituting these expressions into $u^{(1)}$, we get:

$$\begin{split} u^{(1)}(t, \boldsymbol{x}') &\approx \int \nu(\boldsymbol{z})P\left(\left(t - \frac{vx_{1}'}{c} - \frac{r}{c} + \frac{vz_{1}}{c^{2}} - \frac{r - z_{1}\frac{v}{c}}{c} \right)\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right) \\
&= \int\nu(\boldsymbol{z})P\left(\left(t - \frac{vx_{1}'}{c} - \frac{r}{c} + \frac{vz_{1}}{c^{2}} - \frac{r}{c} + z_{1}\frac{v}{c^{2}} \right)\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right) \\
&= \int\nu(\boldsymbol{z})P\left(\left(t - \frac{vx_{1}'}{c} - 2\frac{r}{c} + 2\frac{vz_{1}}{c^{2}} \right)\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right). \end{split}$$

Within the $P(\cdot)$ expression, $x_{1}'$ denotes the first coordinate of the antenna when it receives the signal initially, which is the pulse round-trip time (RTT) of the antenna to the target, from the antenna, and reflected back to the antenna. Therefore, we will say $\boldsymbol{x} = \left(0, -L, H \right)$ and $\boldsymbol{x}' = \left(vt, -L, H \right)$, where the orbit of the antenna is estimated as a straight line. Using the Law of Cosines:

$$\begin{split} t &= \frac{r}{c} + \frac{1}{c}\sqrt{r^{2} + \left(vt \right)^{2} - 2rvt\cos(\gamma_{\boldsymbol{z}})} \\
&= \frac{2r\left(c - v\cos(\gamma_{\boldsymbol{z}}) \right)}{c^{2} - v^{2}} \\
&\approx \frac{2r}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right). \end{split}$$

Therefore,

$$\begin{split} vt &\approx v\left(\frac{2r}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \\
&= 2r\frac{v}{c} - 2r\frac{v^{2}}{c^{2}}\cos(\gamma_{\boldsymbol{z}}), \end{split}$$

and

$$x_{1}' \approx 2r\frac{v}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right).$$

Then, within $P(\cdot)$, $-2\frac{r}{c} - \frac{vx_{1}'}{c^{2}}$ becomes

$$\begin{split} -2\frac{r}{c} - \frac{vx_{1}'}{c^{2}} &= -2\frac{r}{c} - \frac{2r}{c}\frac{v^{2}}{c^{2}}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \\
&= -2\frac{r}{c} - 2\frac{r}{c}\frac{v^{2}}{c^{2}} + 2r\frac{v^{3}}{c^{4}}\cos(\gamma_{\boldsymbol{z}}) \\
&\approx -2\frac{r}{c}\left(1 + \frac{v^{2}}{c^{2}} \right) \\
&\approx -2\frac{r}{c}. \end{split}$$

Now, substituting this expression into $u^{(1)}$ we get:

$$\begin{split} u^{(1)}(t, \boldsymbol{x}') &\approx \int\nu(\boldsymbol{z})P\left(\left(t - 2\frac{r}{c} + 2\frac{vz_{1}}{c^{2}} \right)\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right)d\boldsymbol{z} \\
&= \int\nu(\boldsymbol{z})P\left(\left(t - 2\frac{r}{c}\left(1 - \frac{vz_{1}}{cr} \right)\right)\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right)d\boldsymbol{z} \\
&= \int\nu(\boldsymbol{z})P\left(\left(t - 2\frac{r}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right)\right)\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right)d\boldsymbol{z} \\
&\approx \int\nu(\boldsymbol{z})P\left(t \left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}})\right) - 2\frac{r}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right)\right) \cdot \chi_{\Theta}\left(\frac{z_{1} - \frac{vr}{c}}{R} \right)d\boldsymbol{z}. \end{split}$$

This provides a mathematical model for the SAR raw data while taking into account the physical linear Doppler effect in fast time. We also recall that $r$ is the distance between the emitting location of the antenna $\boldsymbol{x}$ and the target $\boldsymbol{z}$: $r = \left|\boldsymbol{z} - \boldsymbol{x}\right| = R_{\boldsymbol{z}}.$ Where $2\frac{r}{c}$ is the retarded moment of time due to travel delay.

Next, we need to apply the matched filter to obtain the imaging kernel in order to define our image. The matched filter portion will have an overbar over it in order to maximize the filter response of $\nu(\boldsymbol{z})$,

$$I_{\boldsymbol{x}}(\boldsymbol{y}) = \int d\boldsymbol{z}\nu(\boldsymbol{z})\chi_{\Theta}\left(\frac{z_{1} - x_{1}}{R} \right) \cdot \int_{\chi}dt \overline{P\left(t\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}) - 2\frac{R_{\boldsymbol{y}}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}) \right) \right) \right)} \cdot P\left(t\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) - 2\frac{R_{\boldsymbol{z}}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}) \right) \right),$$

where the interior integral is the Point Spread Function (PSF) $W_{\boldsymbol{x}}(\boldsymbol{y}, \boldsymbol{z}).$ The $x_{1}$ argument in $\chi_{\Theta}(\cdot)$ denotes where the radiation signal was emitted from on the antenna, where we will now be looking at multiple emitting locations $\boldsymbol{x}^{n} = (x_{1}^{n}, -L, H).$ The formula for the single-look angles is given by:

$$I_{\boldsymbol{x}^{n}}(\boldsymbol{y}) = \int d\boldsymbol{z}\nu(\boldsymbol{z})\chi_{\Theta}\left(\frac{z_{1} - x_{1}^{n}}{R} \right) \cdot \int_{\chi}dt \overline{P\left(t\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) - 2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) \right) \right)} \cdot P\left(t\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) - 2\frac{R_{\boldsymbol{z}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \right).$$

Where,

$$R_{\boldsymbol{y}}^{n} = \left|\boldsymbol{y} - \boldsymbol{x}^{n} \right|, \quad \cos(\gamma_{\boldsymbol{y}}^{n}) = \frac{y_{1} - x_{1}^{n}}{R_{\boldsymbol{y}}^{n}},$$

and

$$R_{\boldsymbol{z}}^{n} = \left|\boldsymbol{z} - \boldsymbol{x}^{n} \right|, \quad \cos(\gamma_{\boldsymbol{z}}^{n}) = \frac{z_{1} - x_{1}^{n}}{R_{\boldsymbol{z}}^{n}}.$$

Where $\boldsymbol{x}^{n}$ is the sequence of emitting locations of the antenna on the orbit that form the synthetic aperture for a given image point $\boldsymbol{y}.$ Then, defining the overall image using the individual contributions of each $I_{\boldsymbol{x}^{n}}(\boldsymbol{y})$ where there is overlap between the image, $y_{1}$, and target location, $z_{1}$, with reference to the emission locations, $x_{1}^{n}$, we obtain:

$$\begin{split} I(\boldsymbol{y}) &= \sum_{n}\chi_{L_{SA}}\left(y_{1} - x_{1}^{n} \right)I_{\boldsymbol{x}^{n}}(\boldsymbol{y}) \\
&= \int\left[\sum_{n}\chi_{L_{SA}}\left(y_{1} - x_{1}^{n} \right)\chi_{L_{SA}}\left(z_{1} - x_{1}^{n} \right)W_{\boldsymbol{x}^{n}}(\boldsymbol{y}, \boldsymbol{z}) \right]\nu(\boldsymbol{z})d\boldsymbol{z} \\
&= \int W(\boldsymbol{y}, \boldsymbol{z})\nu(\boldsymbol{z})d\boldsymbol{z}. \end{split}$$

Where $W(\boldsymbol{y}, \boldsymbol{z})$ is the imaging kernel, which is a coherent sum of the PSFs, $W_{\boldsymbol{x}^{n}}(\boldsymbol{y}, \boldsymbol{z})$ given by the $dt$ integral, previously. Now, we introduce new notation to help with compactness of the formula:

$$\kappa_{\boldsymbol{y}}^{n} = 1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \quad \textrm{and} \quad \kappa_{\boldsymbol{z}}^{n} = 1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}).$$

Substituting $\kappa_{\boldsymbol{y}}^{n}$ into the $\boldsymbol{y}$ term in $P(\cdot)$,

$$t\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) - 2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) = \left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)\left(t - 2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)^{-1} \right).$$

Since, we know $\gamma_{\boldsymbol{y}}^{n}$ & $\gamma_{\boldsymbol{z}}^{n}$ needs to be as close to $\frac{\pi}{2}$ in order to maximize the response of the kernel, we know that $\cos(\gamma_{\boldsymbol{y}}^{n})$ will be very small and the $cos(\gamma_{\boldsymbol{y}}^{n})$ term also has a factor of $\frac{v}{c}$, where $c \gg v$. Therefore, $2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \ll 1$, and using the property $\frac{1}{1 + x} \simeq 1 - x$ when $x \ll 1$. The multiplication on the right-hand side of the second parenthesis term will become:

$$\begin{split} -2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)^{-1} &\approx -2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)\left(1 - 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) \\
&= -2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 - 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) - 2\frac{v^{2}}{c^{2}}\cos^{2}(\gamma_{\boldsymbol{y}}^{n}) \right) \\
&\approx -2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right), \end{split}$$

where we dropped the $\frac{v^{2}}{c^{2}}$ term. Continuing the substitution into the $\boldsymbol{y}$ term in $P(\cdot)$,

$$\begin{split} t\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) - 2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) &\approx \left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)\left(t - 2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)\right) \\
&= \kappa_{\boldsymbol{y}}^{n}\left(t - t_{\boldsymbol{y}}^{n} \right), \end{split}$$

where

$$t_{\boldsymbol{y}}^{n} = 2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) \quad \textrm{and} \quad t_{\boldsymbol{z}}^{n} = 2\frac{R_{\boldsymbol{z}}^{n}}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right).$$

Where the interrogation of the $\boldsymbol{z}$ term in $P(\cdot)$ shows a similar result. Therefore,

$$I_{\boldsymbol{x}^{n}}(\boldsymbol{y}) = \int d\boldsymbol{z}\nu(\boldsymbol{z})\chi_{\Theta}\left(\frac{z_{1} - x_{1}^{n}}{R} \right) \int_{\chi}dt \overline{P\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right)}P\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right),$$

where

$$W_{\boldsymbol{x}^{n}}(\boldsymbol{y}, \boldsymbol{z}) = \int_{\chi}dt \overline{P\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right)}P\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right).$$

Using the following definition, derived in a previous homework,

$$\begin{split} P(t) &= A(t)e^{-i\omega_{o}t} \\
&= \chi_{\tau}(t)e^{-i\alpha t^{2}}e^{-i\omega_{o}t}, \end{split}$$

we can define $W_{\boldsymbol{x}^{n}}(\boldsymbol{y}, \boldsymbol{z})$ as:

$$\begin{split} W_{\boldsymbol{x}^{n}}(\boldsymbol{y}, \boldsymbol{z}) &= \int_{\chi}dt \overline{A\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right)}A\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right)e^{i\omega_{0}\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n}}e^{-i\omega_{0}\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n}} \\
&= \int_{\chi}dt \chi_{\tau}\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right)e^{i\alpha\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right)^{2}} \chi_{\tau}\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right)e^{-i\alpha\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right)^{2}} e^{i\omega_{0}\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n}}e^{-i\omega_{0}\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n}}. \end{split}$$

Finally, defining the entire imaging kernel, $W(\boldsymbol{y}, \boldsymbol{z})$:

$$\begin{split} W(\boldsymbol{y}, \boldsymbol{z}) &= \sum_{n}\chi_{L_{SA}}(y_{1} - x_{1}^{n})\chi_{L_{SA}}(z_{1} - x_{1}^{n}) \cdot W_{\boldsymbol{x}^{n}}(\boldsymbol{y}, \boldsymbol{z}) \\
&= \sum_{n}\chi_{L_{SA}}(y_{1} - x_{1}^{n})\chi_{L_{SA}}(z_{1} - x_{1}^{n}) \cdot \int_{\chi}dt \chi_{\tau}\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right)e^{i\alpha\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right)^{2}} \chi_{\tau}\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right)e^{-i\alpha\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right)^{2}} e^{i\omega_{0}\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n}}e^{-i\omega_{0}\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n}}. \end{split}$$

Notice the indicator functions, $\chi_{\tau}$, have different arguments in comparison to the Generalized Ambiguity Function (GAF) derived for the non-doppler corrected GAF. This indicates to us the stretching of time that occurs during the RTT of the emitted pulse from the antenna and the reception of the scattered field at the moving antenna. Thus, these intervals that define non-trivial contributions to the imaging kernel will have different lengths between the $\boldsymbol{y}$ and $\boldsymbol{z}$ terms. We define these indicator functions below:

$$\chi_{\tau}\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right) = \begin{cases} 1, & \textrm{if } t \in \left[t_{\boldsymbol{y}}^{n} - \frac{\tau}{2\kappa_{\boldsymbol{y}}^{n}}, t_{\boldsymbol{y}}^{n} + \frac{\tau}{2\kappa_{\boldsymbol{y}}^{n}} \right] \\ 0, & \textrm{otherwise} \end{cases}$$

$$\chi_{\tau}\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right) = \begin{cases} 1, & \textrm{if } t \in \left[t_{\boldsymbol{z}}^{n} - \frac{\tau}{2\kappa_{\boldsymbol{z}}^{n}}, t_{\boldsymbol{z}}^{n} + \frac{\tau}{2\kappa_{\boldsymbol{z}}^{n}} \right] \\ 0, & \textrm{otherwise}. \end{cases}$$

The integration is performed over the intersection of these two intervals, i.e., over $\chi_{\tau}\left(\left(t - t_{\boldsymbol{y}}^{n} \right)\kappa_{\boldsymbol{y}}^{n} \right) \cdot \chi_{\tau}\left(\left(t - t_{\boldsymbol{z}}^{n} \right)\kappa_{\boldsymbol{z}}^{n} \right)$. Which would be either

$$\left[t_{\boldsymbol{y}}^{n} - \frac{\tau}{2\kappa_{\boldsymbol{y}}^{n}}, t_{\boldsymbol{z}}^{n} + \frac{\tau}{2\kappa_{\boldsymbol{z}}^{n}} \right] \approx \left[t_{\boldsymbol{y}}^{n} - \frac{\tau}{2}\left(1 - 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right), t_{\boldsymbol{z}}^{n} + \frac{\tau}{2}\left(1 - 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \right]$$

or

$$\left[t_{\boldsymbol{z}}^{n} - \frac{\tau}{2\kappa_{\boldsymbol{z}}^{n}}, t_{\boldsymbol{y}}^{n} + \frac{\tau}{2\kappa_{\boldsymbol{y}}^{n}} \right] \approx \left[t_{\boldsymbol{z}}^{n} - \frac{\tau}{2}\left(1 - 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right), t_{\boldsymbol{y}}^{n} + \frac{\tau}{2}\left(1 - 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) \right].$$

The analysis is the same for both cases, so we shall continue with the first interval definition, corresponding to $t_{\boldsymbol{z}}^{n} \lt t_{\boldsymbol{y}}^{n}$. We also introduce the following notation:

$$T^{n} = \frac{t_{\boldsymbol{y}}^{n} - t_{\boldsymbol{z}}^{n}}{2}.$$

Performing similar derivations of the range, $W_{R}(\boldsymbol{y}, \boldsymbol{z})$, and azimuthal, $W_{\sum}(\boldsymbol{y}, \boldsymbol{z})$, factors of $W(\boldsymbol{y}, \boldsymbol{z})$ as the case without the doppler corrections, we get the following expressions:

$$W_{\sum}(\boldsymbol{y}, \boldsymbol{z}) = \sum_{n}\chi_{L_{SA}}(y_{1} - x_{1}^{n})\chi_{L_{SA}}(z_{1} - x_{1}^{n})e^{-i\omega_{0}T^{n}\left(\kappa_{\boldsymbol{y}}^{n} + \kappa_{\boldsymbol{z}}^{n} \right)} $$

and

$$W_{R}(\boldsymbol{y}, \boldsymbol{z}) = \tau^{c}sinc\left(\tau^{c}\left(\alpha T^{c}\left(\left(\kappa_{\boldsymbol{y}}^{c} \right)^{2} + \left(\kappa_{\boldsymbol{z}}^{c} \right)^{2} \right) - \omega_{0}\frac{\left(\kappa_{\boldsymbol{y}}^{c} - \kappa_{\boldsymbol{z}}^{c} \right)}{2} \right) \right),$$

where $\tau^{c}$ is the center values of the slowly varying envelope portion of the range factor, so all center values would be constants and taken outside of the summation with respect to n. Solving for $T^{n}\left(\kappa_{\boldsymbol{y}}^{n} + \kappa_{\boldsymbol{z}}^{n} \right)$ using the previously defined expressions that we will rewrite here for ease of use:

$$\kappa_{\boldsymbol{y}}^{n} = 1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \quad \textrm{and} \quad \kappa_{\boldsymbol{z}}^{n} = 1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}),$$

$$T^{n} = \frac{t_{\boldsymbol{y}}^{n} - t_{\boldsymbol{z}}^{n}}{2},$$

$$t_{\boldsymbol{y}}^{n} = 2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) \quad \textrm{and} \quad t_{\boldsymbol{z}}^{n} = 2\frac{R_{\boldsymbol{z}}^{n}}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right).$$

$$\begin{split} T^{n}\left(\kappa_{\boldsymbol{y}}^{n} + \kappa_{\boldsymbol{z}}^{n} \right) &= \frac{t_{\boldsymbol{y}}^{n} - t_{\boldsymbol{z}}^{n}}{2}\left(1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) + 1 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \\
&= \frac{t_{\boldsymbol{y}}^{n} - t_{\boldsymbol{z}}^{n}}{2}\left(2 + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) + 2\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \\
&= \left(t_{\boldsymbol{y}}^{n} - t_{\boldsymbol{z}}^{n} \right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \\
&= \left(2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) - 2\frac{R_{\boldsymbol{z}}^{n}}{c}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right)\right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \\
&= \frac{2}{c}\left(R_{\boldsymbol{y}}^{n}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) - R_{\boldsymbol{z}}^{n}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right)\right)\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right). \end{split}$$

Just analyzing the $R_{\boldsymbol{y}}^{n}$ term, we get:

$$\begin{split} & \frac{2}{c}\left(R_{\boldsymbol{y}}^{n}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) + R_{\boldsymbol{y}}^{n}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)\left(\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) + R_{\boldsymbol{y}}^{n}\left(1 - \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right)\left(\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \right) \\
& \frac{2}{c}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{y}}^{n}\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) + R_{\boldsymbol{y}}^{n}\left(\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) - \frac{v^{2}}{c^{2}}\cos^{2}(\gamma_{\boldsymbol{y}}^{n}) \right) + R_{\boldsymbol{y}}^{n}\left(\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) - \frac{v^{2}}{c^{2}}\cos(\gamma_{\boldsymbol{y}}^{n})\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \right) \\
& \frac{2}{c}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{y}}^{n}\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) + R_{\boldsymbol{y}}^{n}\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) + R_{\boldsymbol{y}}^{n}\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \\
& \frac{2}{c}\left(R_{\boldsymbol{y}}^{n} + R_{\boldsymbol{y}}^{n}\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) \\
& 2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right), \end{split}$$

where we neglected the $\frac{v^{2}}{c^{2}}$ terms. We can easily see the extension of this analysis to the $R_{\boldsymbol{z}}^{n}$ term:

$$-2\frac{R_{\boldsymbol{z}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right).$$

Consequently,

$$\begin{split} T^{n}(\kappa_{\boldsymbol{y}}^{n} + \kappa_{\boldsymbol{z}}^{n}) &\approx 2\frac{R_{\boldsymbol{y}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) -2\frac{R_{\boldsymbol{z}}^{n}}{c}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) \\
&= \frac{2}{c}\left(R_{\boldsymbol{y}}^{n}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) \right) - R_{\boldsymbol{z}}^{n}\left(1 + \frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) \right) \\
&= \frac{2}{c}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n} + R_{\boldsymbol{y}}^{n}\frac{v}{c}\cos(\gamma_{\boldsymbol{z}}^{n}) - R_{\boldsymbol{z}}^{n}\frac{v}{c}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) \\
&= \frac{2}{c}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n} + \frac{v}{c}\left(R_{\boldsymbol{y}}^{n}\cos(\gamma_{\boldsymbol{z}}^{n}) - R_{\boldsymbol{z}}^{n}\cos(\gamma_{\boldsymbol{y}}^{n}) \right) \right) \\
&= \frac{2}{c}\left(R_{\boldsymbol{y}}^{n} - R_{\boldsymbol{z}}^{n} + \frac{v}{c}\left(\frac{R_{\boldsymbol{y}}^{n}}{R_{\boldsymbol{z}}^{n}}\left(z_{1} - x_{1}^{n} \right) - \frac{R_{\boldsymbol{z}}^{n}}{R_{\boldsymbol{y}}^{n}}\left(y_{1} - x_{1}^{n} \right) \right) \right). \end{split}$$

Where $R_{Dop} = \frac{R_{\boldsymbol{y}}^{n}}{R_{\boldsymbol{z}}^{n}}\left(z_{1} - x_{1}^{n} \right) - \frac{R_{\boldsymbol{z}}^{n}}{R_{\boldsymbol{y}}^{n}}\left(y_{1} - x_{1}^{n} \right)$. Using the Pythagorean expressions for the travel distances,

$$\begin{split} \left(R_{\boldsymbol{y}}^{n} \right)^{2} &= R'^{2} + \left(x_{1}^{n} - y_{1} \right)^{2} \\
\left(R_{\boldsymbol{z}}^{n} \right)^{2} &= R'^{2} + \left(x_{1}^{n} - z_{1} \right)^{2}, \end{split}$$

where

$$\begin{split} R'^{2} &= R^{2} + l^{2} - 2Rl\cos(\frac{\pi}{2} + \theta) \\
&= R^{2} + l^{2} + 2Rl\sin(\theta) \\
&\approx \left(R + l\sin(\theta) \right)^{2} \end{split}$$

is given by the cosine law. $R$ is the slant range, slant distance from the antenna to the target. $\theta$ is the angle of incidence of the antenna when emitting the pulse. $l$ is the difference in the range between the image and target, i.e., $y_{2} - z_{2}$. $R_{Dop}$, defined above, describes the linear effect with respect to $\frac{v}{c}$ of the antenna motion on the azimuthal factor of the GAF. Additionally, introducing the following definitions:

$$\tilde{x} = \tilde{n}\Delta x_{1} \quad and \quad d = \frac{y_{1} - z_{1}}{2},$$

such that

$$y_{1} - x_{1}^{n} = d - \tilde{x} \quad and \quad z_{1} - x_{1}^{n} = -(d + \tilde{x}).$$

Substituting these values into $R_{Dop}$ and Tayloring the necessary expressions,

$$\begin{split} R_{Dop} &= \frac{\sqrt{R'^{2} + \left(x_{1}^{n} - y_{1} \right)^{2}}}{\sqrt{R'^{2} + \left(x_{1}^{n} - z_{1} \right)^{2}}}\frac{z_{1} - x_{1}^{n}}{\left|\boldsymbol{z} - \boldsymbol{x}^{n} \right|} - \frac{\sqrt{R'^{2} + \left(x_{1}^{n} - z_{1} \right)^{2}}}{\sqrt{R'^{2} + \left(x_{1}^{n} - y_{1} \right)^{2}}}\frac{y_{1} - x_{1}^{n}}{\left|\boldsymbol{y} - \boldsymbol{x}^{n} \right|} \\
&= -\frac{d + \tilde{x}}{\sqrt{R'^{2} + \left(d + \tilde{x} \right)^{2}}}\frac{\sqrt{R'^{2} + \left(\tilde{x} - d \right)^{2}}}{R} - \frac{d - \tilde{x}}{\sqrt{R'^{2} + \left(\tilde{x} - d \right)^{2}}}\frac{\sqrt{R'^{2} + \left(d + \tilde{x} \right)^{2}}}{R} \\
&= -\frac{d + \tilde{x}}{\sqrt{R^{2} + \left(d + \tilde{x} \right)^{2}}}\sqrt{R'^{2} + \left(\tilde{x} - d \right)^{2}} - \frac{d - \tilde{x}}{\sqrt{R'^{2} + \left(\tilde{x} - d \right)^{2}}}\sqrt{R^{2} + \left(d + \tilde{x} \right)^{2}} \\
&\approx -\frac{d + \tilde{x}}{R\left(1 + \frac{1}{2} \frac{\left(d + \tilde{x} \right)^{2}}{R^{2}}\right)}R'\left(1 + \frac{1}{2}\frac{\left(\tilde{x} - d \right)^{2}}{R'^{2}} \right) - \frac{\left(d - \tilde{x}\right)^{2}}{R'\left(1 + \frac{1}{2}\frac{\tilde{x} - d}{R'^{2}} \right)}R\left(1 + \frac{1}{2}\frac{\left(d + \tilde{x} \right)^{2}}{R^{2}} \right) \\
&= -\frac{d + \tilde{x}}{R + \frac{1}{2} \frac{\left(d + \tilde{x} \right)^{2}}{R}}\left(R' + \frac{1}{2}\frac{\left(\tilde{x} - d \right)^{2}}{R'}\right) - \frac{d - \tilde{x}}{R' + \frac{1}{2}\frac{\left(\tilde{x} - d\right)^{2}}{R'}}\left(R + \frac{1}{2}\frac{\left(d + \tilde{x} \right)^{2}}{R}\right). \end{split}$$

Inspecting the first term and taking the derivative with respect to $\tilde{x}$,

$$-\frac{d + \tilde{x}}{R + \frac{1}{2} \frac{\left(d + \tilde{x} \right)^{2}}{R}}\left(R' + \frac{1}{2}\frac{\left(\tilde{x} - d \right)^{2}}{R'}\right) = -\frac{dR'}{R + \frac{1}{2}\frac{\left(d + \tilde{x} \right)^{2}}{R}} + \frac{\frac{d}{2}\frac{\left(\tilde{x} - d \right)^{2}}{R'}}{R + \frac{1}{2}\frac{\left(d + \tilde{x} \right)^{2}}{R}} + \frac{\tilde{x}R'}{R + \frac{1}{2}\frac{\left(d + \tilde{x} \right)^{2}}{R}} + \frac{\frac{\tilde{x}}{2}\frac{\left(\tilde{x} - d \right)^{2}}{R'}}{R + \frac{1}{2}\frac{\left(d + \tilde{x} \right)^{2}}{R}}.$$

Where we will denote:

$$\begin{split} R_{Dop_{1}}^{(1)} &= -\frac{dR'}{R + \frac{1}{2}\frac{\left(d + \tilde{x} \right)^{2}}{R}} \\
R_{Dop_{1}}^{(2)} &= \frac{\frac{d}{2}\frac{\left(\tilde{x} - d \right)^{2}}{R'}}{R + \frac{1}{2}\frac{\left(d + \tilde{x} \right)^{2}}{R}} \\
R_{Dop_{1}}^{(3)} &= \frac{\tilde{x}R'}{R + \frac{1}{2}\frac{\left(d + \tilde{x} \right)^{2}}{R}} \\
R_{Dop_{1}}^{(4)} &= \frac{\frac{\tilde{x}}{2}\frac{\left(\tilde{x} - d \right)^{2}}{R'}}{R + \frac{1}{2}\frac{\left(d + \tilde{x} \right)^{2}}{R}}. \end{split}$$

Solving for $\frac{\partial R_{Dop_{1}}^{(i)}}{\partial\tilde{x}}\bigg\rvert_{\tilde{x}=d=0}, i \in [1, 2, 3, 4]$ where $\Xi$ will denote don't care values, indicating an expression that has no impact on the result due to the fact that $\tilde{x} = d =0$ (we are also able to easily see that there will never be a divide by zero occurrence due to $R^{2}$ or $R'^{2}$ always being present in each denominator):

$$\begin{split} \frac{\partial R_{Dop_{1}}^{(1)}}{\partial\tilde{x}}\bigg\rvert_{\tilde{x}=d=0} &= \frac{\frac{d\tilde{x}}{R} + \frac{\tilde{x}}{R}\Xi}{\Xi} \\
&= \frac{0 + 0\Xi}{\Xi} \\
&= 0, \\
\\
\frac{\partial R_{Dop_{1}}^{(2)}}{\partial\tilde{x}}\bigg\rvert_{\tilde{x}=d=0} &= \frac{\frac{d\left(\tilde{x} - d \right)}{R'}\left(R + \frac{1}{2}\frac{\left(d + \tilde{x} \right)^{2}}{R} \right) - \frac{d}{2}\frac{\left(\tilde{x} - d \right)^{2}}{R'}\left(\frac{\frac{d\tilde{x}}{R} + \frac{\tilde{x}}{R}\Xi }{R + \frac{1}{2}\left(\frac{d^{2} + 2d\tilde{x} + \tilde{x}^{2}}{R} \right)^{2}} \right) }{\Xi} \\
&= \frac{0 - 0}{\Xi} \\
&= 0, \\
\\
\frac{\partial R_{Dop_{1}}^{(3)}}{\partial\tilde{x}}\bigg\rvert_{\tilde{x}=d=0} &= \frac{R'\left(R + \frac{1}{2}\frac{\left(d + \tilde{x} \right)^{2}}{R} \right) - \tilde{x}R\Xi }{\left(R + \frac{1}{2}\frac{\left(d + \tilde{x} \right)^{2}}{R} \right)^{2}} \\
&= \frac{R'R}{R^{2} + \frac{\left(d + \tilde{x} \right)^{2}}{R} + \frac{1}{4}\frac{\left(d + \tilde{x} \right)^{4}}{R^{2}}} \\
&= \frac{R'R}{R^{2}} \\
&= \frac{R'}{R}, \\
\\
\frac{\partial R_{Dop_{1}}^{(4)}}{\partial\tilde{x}}\bigg\rvert_{\tilde{x}=d=0} &= \frac{0}{\Xi} \\
&= 0. \end{split}$$

Therefore,

$$\frac{\partial R_{Dop_{1}}}{\partial\tilde{x}}\bigg\rvert_{\tilde{x}=d=0} = \frac{R'}{R}.$$

It is easy to see the extension of this:

$$\frac{\partial R_{Dop_{2}}}{\partial\tilde{x}}\bigg\rvert_{\tilde{x}=d=0} = -\frac{R}{R'}.$$

Where,

$$\frac{\partial R_{Dop}}{\partial\tilde{x}}\bigg\rvert_{\tilde{x}=d=0} = \frac{R'}{R} - \frac{R}{R'} \approx -2\frac{l\sin(\theta)}{R},$$

and, therefore,

$$R_{Dop} \approx -2d - 2\frac{l\sin(\theta)}{R}\tilde{x}.$$

&= \frac{\left(vz_{1} - c^{2}t \right) \pm \sqrt{\left(c^{2}t - vz_{1} \right)^{2} -\left(v^{2} - c^{2} \right)\left(z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2} - \left(ct \right)^{2} \right) }}{\left(v^{2} - c^{2} \right)}. \end{split}$$

Analyzing the second term under the square root:

$$\begin{split} \left(v^{2} - c^{2} \right)\left(z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2} - \left(ct \right)^{2} \right) &= z_{1}^{2}v^{2} + v^{2}\left(z_{2} + L \right)^{2} + v^{2}\left(z_{3} - H \right)^{2} - v^{2}c^{2}t^{2} - c^{2}z_{1}^{2} - c^{2}\left(z_{2} + L \right)^{2} - \left(z_{3} - H \right)^{2} - c^{4}t^{2} \\
&= v^{2}\left(z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2} - c^{2}t^{2} \right) - c^{2}\left(z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2} - c^{2}t^{2} \right) \\
&= c^{2}v^{2}\left(\frac{z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2} - c^{2}t^{2}}{c^{2}} - \frac{z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2} - c^{2}t^{2}}{v^{2}} \right). \end{split}$$

Analyzing the first term under the square root:

$$\left(c^{2}t - vz_{1} \right)^{2} = c^{2}\left(t - \frac{vz_{1}}{c^{2}} \right).$$

Analyzing the square root expression:

$$\begin{split} &= \sqrt{c^{4}\left(t - \frac{vz_{1}}{c^{2}} \right)^{2} - c^{2}v^{2}\left(\frac{z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2} - c^{2}t^{2}}{c^{2}} - \frac{z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2} - c^{2}t^{2}}{v^{2}} \right)} \\
&= c^{2}\sqrt{\left(t - \frac{vz_{1}}{c^{2}} \right)^{2} - \frac{v^{2}}{c^{2}}\left(\frac{z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2}}{c^{2}} - t^{2} - \frac{z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2} - c^{2}t^{2}}{v^{2}} \right)} \\
&= c^{2}\sqrt{\left(t - \frac{vz_{1}}{c^{2}} \right)^{2} - \left(\frac{v^{2}}{c^{2}} - 1 \right)\left(\frac{z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2}}{c^{2}} - t^{2} \right)} \\
\\
\beta &= \sqrt{1 - \frac{v^{2}}{c^{2}}} \\
\beta^{2} &= 1 - \frac{v^{2}}{c^{2}} \\
\\
&= c^{2}\sqrt{\left(t - \frac{vz_{1}}{c^{2}} \right)^{2} - \beta^{2}\left(t^{2} - \frac{z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2}}{c^{2}} \right)}. \end{split}$$

Substituting this back into the solution for $t'$:

$$\begin{split} t' &= \frac{\left(vz_{1} - c^{2}t \right) \pm c^{2}\sqrt{\left(t - \frac{vz_{1}}{c^{2}} \right)^{2} - \beta^{2}\left(t^{2} - \frac{z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2}}{c^{2}} \right) }}{v^{2} - c^{2}} \\
&= \frac{-c^{2}\left(t - \frac{vz_{1}}{c^{2}} \right) \pm c^{2}\sqrt{\left(t - \frac{vz_{1}}{c^{2}} \right)^{2} - \beta^{2}\left(t^{2} - \frac{z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2}}{c^{2}} \right) }}{v^{2} - c^{2}} \\
&= -\frac{c^{2}}{v^{2}}\left(\frac{t - \frac{vz_{1}}{c^{2}} \mp \sqrt{\left(t - \frac{vz_{1}}{c^{2}} \right)^{2} - \beta^{2}\left(t^{2} - \frac{z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2}}{c^{2}} \right) } }{\frac{c^{2}}{v^{2}} - 1} \right) \\
&= \frac{1}{\beta^{2}}\left(t - \frac{vz_{1}}{c^{2}} \mp \sqrt{\left(t - \frac{vz_{1}}{c^{2}} \right)^{2} - \beta^{2}\left(t^{2} - \frac{z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2}}{c^{2}} \right) } \right) \\
\\
\sigma &= \frac{1}{\beta}\left(t - \frac{vz_{1}}{c^{2}} \right) \\
\\
&= \frac{\sigma}{\beta} \pm \frac{1}{\beta c}\sqrt{\frac{\left(z_{1} - vt \right)^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2} }{\beta^{2}}} \\
\\
\zeta_{1} &= \frac{1}{\beta}\left(-vt + z_{1} \right) \\
\\
&= \frac{\sigma}{\beta} \pm \frac{\zeta_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2}}{\beta c} \\
\\
\rho &= \sqrt{\zeta_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2}}
\\
\\
&= \frac{\sigma}{\beta} \pm \frac{\rho}{\beta c}. \end{split}$$

Where we then choose the solution corresponding to $t' = \frac{\sigma}{\beta} - \frac{\rho}{\beta c}$. We choose this solution, because we are solving for the time at which the scattered signal will be received at the antenna. If we were to choose the $+$ root, we would be looking at some time that would not align with the travel time of the signal and thus only exist at a negative time in order to satisfy $\mu(t') = ct$. Now to solve for the denominator:

$$\begin{split} c\left|\boldsymbol{z} - \boldsymbol{x}(t') \right| - v\left(z_{1} - vt' \right) &= c^{2}\left(t - t' \right) - v\left(z_{1} - vt' \right) \\
&= c^{2}t - z_{1}v - t'\left(c^{2} - v^{2} \right) \\
&= c^{2}\left(t - \frac{z_{1}v}{c^{2}} - t'\left(1 - \frac{v^{2}}{c^{2}} \right) \right) \\
&= c^{2}\left(t - \frac{z_{1}v}{c^{2}} - t'\beta^{2} \right) \\
\\
&\textrm{using the solution for t'} \\
\\
&= c^{2}\sqrt{\left(t - \frac{vz_{1}}{c^{2}} \right)^{2} - \beta^{2}\left(t^{2} - \frac{z_{1}^{2} + \left(z_{2} + L \right)^{2} + \left(z_{3} - H \right)^{2}}{c^{2}} \right)} \\
&= \beta c \rho. \end{split}$$

Substituting our solutions for $t'$ and the denominator back into $u(t, \boldsymbol{z})$:

$$\begin{split} u(t, \boldsymbol{z}) &= \frac{1}{4\pi}\frac{cP\left(t' \right)}{c\left|\boldsymbol{z} - \boldsymbol{x}(t') \right| - v\left(z_{1} - vt' \right)}\bigg\rvert_{\mu(t') = ct} \\
&= \frac{1}{4\pi}\frac{cP\left(\frac{\sigma}{\beta} - \frac{\rho}{\beta c} \right)}{\beta c \rho} \\
&= \frac{1}{4\pi\beta}\frac{P\left(\frac{1}{\beta}\left(\sigma - \frac{\rho}{c} \right) \right)}{\rho}. \end{split}$$

Where we can see that $u(t, \boldsymbol{z})$ does indeed reduce down to our original statement at the beginning of this problem.
